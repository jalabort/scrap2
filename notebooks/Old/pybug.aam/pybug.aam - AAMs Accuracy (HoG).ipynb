Test pybug.aam.AAM fitting accuracy

```python
import menpo.io as pio
from menpo.landmark import labeller, ibug_68_trimesh

training_images = []
# load landmarked imarges
for i in pio.import_images('/data/PhD/DataBases/lfpw/trainset/*.png',
                     max_images=811):
    i.crop_to_landmarks_proportion(0.1)
    if i.n_channels == 3:
        training_images.append(i.as_greyscale(mode='luminosity'))
    else:
        training_images.append(i)
```

```python
%matplotlib inline
#visualize the first image
training_images[0].landmarks['PTS'].view_new()
```

```python
from menpo.fitmultilevel.aam import AAMBuilder


# build aam
aam = AAMBuilder(feature_type=None, diagonal_range=100).build(training_images, group='PTS')
```

```python
import numpy as np

# low resolution level
%matplotlib inline
aam.instance(level=0).view() 
# middle resolution level
aam.instance(level=1).view_new()
# highest resolution level
aam.instance().view_new()
```

```python
test_images = []
# load landmarked imarges
for i in pio.import_images('/data/PhD/DataBases/lfpw/testset/*.png',
                     max_images=22):
    i.crop_to_landmarks_proportion(0.3)
    if i.n_channels == 3:
        test_images.append(i.as_greyscale(mode='luminosity'))
    else:
        test_images.append(i)
```

```python
from menpo.fitmultilevel.aam import LucasKanadeAAMFitter
from menpo.fit.lucaskanade.appearance import (AlternatingForwardCompositional, 
                                              AlternatingForwardAdditive, 
                                              AlternatingInverseCompositional,
                                              SimultaneousForwardCompositional, 
                                              SimultaneousForwardAdditive, 
                                              SimultaneousInverseCompositional,
                                              ProbabilisticForwardCompositional, 
                                              ProbabilisticForwardAdditive, 
                                              ProbabilisticInverseCompositional,
                                              ProjectOutForwardCompositional, 
                                              ProjectOutForwardAdditive, 
                                              ProjectOutInverseCompositional,
                                              AdaptiveForwardCompositional, 
                                              AdaptiveForwardAdditive, 
                                              AdaptiveInverseCompositional)
from menpo.fit.lucaskanade import FeatureSSD, SSD, ECC, GradientSSD
from menpo.fitmultilevel.featurefunctions import sparse_hog
                                              

fitter = LucasKanadeAAMFitter(aam, algorithm=AlternatingForwardCompositional, residual=GradientSSD,
                              n_shape=[3, 6, 12], n_appearance=100, feature_type='igo')
```

```python
fitting = fitter.fit(test_images[21], noise_std=0, group='PTS')
```

```python
%matplotlib inline
fitting.view_initialization()
fitting.view_final_fitting(new_figure=True)
```
