```python
import pybug.io as pio
breakingbad = pio.import_builtin_asset('breakingbad.jpg')
breakingbad = breakingbad.as_greyscale(mode='luminosity')
breakingbad.crop_to_landmarks_proportion(0.3, group='PTS')

%matplotlib inline
breakingbad.landmarks['PTS'].view()
```

```python
import pybug.io as pio
einstein = pio.import_builtin_asset('einstein.jpg')
einstein.crop_to_landmarks_proportion(0.3, group='PTS')
einstein.constrain_mask_to_landmarks()

%matplotlib inline
einstein.landmarks['PTS'].view()
```

```python
hog_breakingbad = breakingbad.features.hog()

%matplotlib inline
hog_breakingbad.landmarks['PTS'].view(channels=0)
```

```python
from pybug.transform.piecewiseaffine import PiecewiseAffineTransform

transform = PiecewiseAffineTransform(einstein.landmarks['PTS'].lms, breakingbad.landmarks['PTS'].lms)
```

```python
%timeit intensity_warped = breakingbad.warp_to(einstein.mask, transform)
```

```python
%timeit hog_warped = hog_breakingbad.warp_to(einstein.mask, transform)
```
