```python
import menpo.io as pio
from menpo.landmark import labeller, ibug_68_points, ibug_68_trimesh, ibug_49_points

training_images = []
# load landmarked imarges
for i in pio.import_images('/data/PhD/DataBases/lfpw/trainset/*.png', max_images=811):
    i.crop_to_landmarks_proportion(0.4)
    if i.n_channels == 3:
        training_images.append(i.as_greyscale(mode='luminosity'))
    else:
        training_images.append(i)
```

```python
# visualize the first image
%matplotlib inline
training_images[0].landmarks['PTS'].view()
```

```python
image = training_images[0]

def hog_closure(image):
    return image.features.hog(mode='sparse')

%matplotlib inline
a = hog_closure(training_images[0])
a.landmarks['PTS'].view(channels=0)
```

```python
from pybug.aam.base import aam_builder
from pybug.transform.piecewiseaffine import PiecewiseAffineTransform
from pybug.image import Image

# label images
labeller(training_images, 'PTS', ibug_68_trimesh);

# build aam
aam = aam_builder(training_images, 
                  group='PTS', 
                  trilist=training_images[0].landmarks['ibug_68_trimesh'].lms.trilist, 
                  n_levels=3, 
                  downscale=1.1, 
                  feature_type=hog_closure, 
                  max_shape_components=25, 
                  max_appearance_components=250)
```

```python
%matplotlib inline
a.build_mask_around_landmarks((5, 5), group='PTS')
a.landmarks['PTS'].view(channels=0)
```

```python
from sklearn import svm
from sklearn import linear_model 
from pybug.asm.base import asm_builder

def classifier_closure(X, y):
    clf1 = svm.LinearSVC(class_weight='auto')
    clf1.fit(X, y)
    y_pred = clf1.decision_function(X)
    clf2 = linear_model.LogisticRegression(class_weight='auto')
    clf2.fit(y_pred[..., None], y)
    return clf1, clf2

# build aam
asm = asm_builder(training_images, group='PTS', patch_size=(5, 5), n_levels=3,
                  downscale=1.1, scaled_shape_models=False, feature_type=hog_closure, 
                  classifier_type=classifier_closure, max_shape_components=25)
```

```python
del training_images
```

```python
from pybug.asam.base import ASAM

asam = ASAM(asm.shape_models, aam.appearance_models, aam.transform_cls,
            asm.classifiers, hog_closure, asm.reference_shape, asm.downscale,
            asm.patch_size, aam.interpolator)

asam.scaled_shape_models = True
```

```python
# fit test images
asam_fittings = asam_fitter.fit_images(test_images, group='PTS', noise_std=0.02, max_iters=50)
```

```python
from __future__ import division
from pybug.visualize.base import GraphPlotter

def final_error_dist(final_error):
    return _error_dist(final_error)

def initial_error_dist(initial_error):
    return _error_dist(initial_error)

def _error_dist(error):
    n_errors = np.product(error.shape)
    x_axis = np.arange(0, asam_fittings._error_stop, asam_fittings._error_step)
    ed = np.array([np.count_nonzero((limit-asam_fittings._error_step) <= error[error <= limit])
                   for limit in x_axis]) / n_errors
    return ed, x_axis

def final_cumulative_error_dist(final_error):
    ed, x_axis = final_error_dist(final_error)
    ced = _cumulative_error_dist(ed)
    return ced, x_axis

def initial_cumulative_error_dist(initial_error):
    ed, x_axis = initial_error_dist(initial_error)
    ced = _cumulative_error_dist(ed)
    return ced, x_axis

def _cumulative_error_dist(error_dist):
    return np.array([np.sum(error_dist[:j])
                     for j, _ in enumerate(error_dist)])

def plot_error_dist(final_error, initial_error, figure_id=None, new_figure=False, **kwargs):
    title = 'Error Distribution'
    ed, x_axis = final_error_dist(final_error)
    ed = [ed, initial_error_dist(initial_error)[0]]
    return _plot_dist(final_error, initial_error, x_axis, ed, title, figure_id=figure_id,
                                     new_figure=new_figure, y_limit=np.max(ed),
                                     **kwargs)

def plot_cumulative_error_dist(final_error, initial_error, figure_id=None, new_figure=False, **kwargs):
    title = 'Cumulative Error Distribution'
    ced, x_axis = final_cumulative_error_dist(final_error)
    ced = [ced, initial_cumulative_error_dist(initial_error)[0]]
    return _plot_dist(final_error, initial_error, x_axis, ced, title, figure_id=figure_id,
                                     new_figure=new_figure, **kwargs)

def _plot_dist(final_error, initial_error, x_axis, y_axis, title, figure_id=None,
               new_figure=False, y_limit=1, **kwargs):
    legend = [asam_fittings.algorithm +
              '\nmean: {0:.4f}'.format(np.mean(final_error)) +
              'std: {0:.4f}, '.format(np.std(final_error)) +
              'median: {0:.4f}, '.format(np.median(final_error)) +
              'convergence: {0:.2f}, '.format(convergence),
              'Initialization' +
              '\nmean: {0:.4f}, '.format(np.mean(initial_error)) +
              'std: {0:.4f}, '.format(np.std(initial_error)) +
              'median: {0:.4f}, '.format(np.median(initial_error))]
    x_label = asam_fittings._error_text
    y_label = 'Proportion of images'
    axis_limits = [0, 0.05, 0, y_limit]
    return GraphPlotter(figure_id, new_figure, x_axis, y_axis,
                        title=title, legend=legend,
                        x_label=x_label, y_label=y_label,
                        axis_limits=axis_limits).render(**kwargs)
```
