```python
import pybug.io as pio
from pybug.landmark import labeller, ibug_68_points, ibug_68_trimesh

training_images = []
# load landmarked imarges
for i in pio.import_images('/data/PhD/DataBases/lfpw/trainset/*.png'):
    i.crop_to_landmarks_proportion(0.2)
    if i.n_channels == 3:
        training_images.append(i.as_greyscale(mode='luminosity'))
    else:
        training_images.append(i)
```

```python
%matplotlib inline
#visualize the first image
training_images[0].landmarks['PTS'].view()
```

### - Normalize face size and generate gaussian levels:

```python
from pybug.aam.functions import mean_pointcloud
    
group = 'PTS'
label = 'all'
diagonal_range = None
n_levels = 3
downscale = 2
interpolator = 'scipy'

print '- Rescaling images'
training_shapes = [i.landmarks[group][label].lms for i in training_images]
reference_shape = mean_pointcloud(training_shapes)
if diagonal_range:
    x, y = reference_shape.range()
    scale = diagonal_range / np.sqrt(x**2 + y**2)
    Scale(scale, reference_shape.n_dims).apply_inplace(reference_shape)
training_images = [i.rescale_to_reference_shape(reference_shape, group=group, 
                                                label=label, interpolator=interpolator)
                   for i in training_images]


print '- Setting gaussian pyramid generators'
generator = [i.gaussian_pyramid(n_levels=n_levels,
                                downscale=downscale)
             for i in training_images]
```

### - Generate first level

```python
from pybug.aam.functions import compute_features

def hog_closure(image):
    return image.features.hog(cell_size=4, window_step_horizontal=1, window_step_vertical=1)

print '  - Computing feature_type'
training_images = [compute_features(g.next(), hog_closure) for g in generator]
# extract potentially rescaled shapes
training_shapes = [i.landmarks[group][label].lms for i in training_images]
```

```python
import numpy as np
from sklearn import svm
from sklearn import linear_model 
from pybug.asm.functions import build_sampling_grid, get_pos_neg_grid_positions
```

```python
sampling_grid = build_sampling_grid((16, 16))
mask = np.ones(sampling_grid.shape[:-1]).astype(bool)

classifiers = []
for k in np.arange(2):
    
    positive_labels = []
    negative_labels = []
    positive_samples = []
    negative_samples = []
    
    for i, s in zip(training_images, training_shapes):
        
        max_x = i.shape[0] - 1
        max_y = i.shape[1] - 1
            
        point = (np.round(s.points[k, :])).astype(int)
        patch_grid = sampling_grid + point[None, None, ...]
        positive, negative = get_pos_neg_grid_positions(patch_grid, positive_grid_size=(2, 2))
        
        for p in s.points:
            if np.all(p != s.points[k, :]):
                p = (np.round(p)).astype(int)
                neg = sampling_grid + p[None, None, ...]
                negative = np.vstack((negative, neg[mask]))
                
        positive_sample = i.pixels[positive[:, 0], positive[:, 1], :]
        positive_samples.append(positive_sample)
        positive_labels.append(np.ones(positive_sample.shape[0]))
        
        x = negative[:, 0]
        y = negative[:, 1]
        x[x > max_x] = max_x
        y[y > max_y] = max_y
        x[x < 0] = 0
        y[y < 0] = 0
        
        negative_sample = i.pixels[x, y, :]
        print negative_sample.shape
        negative_samples.append(negative_sample)
        negative_labels.append(-np.ones(negative_sample.shape[0]))
    
    positive_samples = np.asanyarray(positive_samples)
    positive_samples = np.reshape(positive_samples, (-1, positive_samples.shape[-1]))
    positive_labels = np.asanyarray(positive_labels).flatten()

    negative_samples = np.asanyarray(negative_samples)
    negative_samples = np.reshape(negative_samples, (-1, negative_samples.shape[-1]))
    negative_labels = np.asanyarray(negative_labels).flatten()
    
    X = np.vstack((positive_samples, negative_samples))
    y = np.hstack((positive_labels, negative_labels))
    
    #svc = svm.SVC(kernel='linear', class_weight='auto', probability=True)
    #svc.fit(X, y)
    #classifiers.append(svc)
    #svc = svm.SVC(kernel='sigmoid', class_weight='auto', probability=True)
    #svc.fit(X, y)
    #classifiers.append(svc)
    #svc = svm.SVC(kernel='rbf', class_weight='auto', probability=True)
    #svc.fit(X, y)
    #classifiers.append(svc)
    
    X = (X - np.mean(X, axis=1)) / np.std(X, axis=1)
    
    svc = svm.LinearSVC(class_weight='auto')
    svc.fit(X, y)
    y_pred = svc.decision_function(X)
    
    #svc = linear_model.SGDClassifier(class_weight='auto')
    #svc.fit(X, y)
    #y_pred = svc.decision_function(X)
    
    lr = linear_model.LogisticRegression(class_weight='auto')
    lr.fit(y_pred[..., None], y)
    classifiers.append((svc, lr))
    
```
