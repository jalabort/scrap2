```python
import pybug.io as pio

training_images = []
# load landmarked imarges
for i in pio.import_images('/data/eyes/Annotated_Flip/Open/Right/*.png'):
    i.crop_to_landmarks_proportion(0.1)
    if i.n_channels == 3:
        training_images.append(i.as_greyscale(mode='luminosity'))
    else:
        training_images.append(i)
```

```python
%matplotlib inline
#visualize the first image
training_images[0].landmarks['PTS'].view_new()
```

```python
from pybug.landmark import \
labeller, ibug_open_eye_points, ibug_open_eye_trimesh, ibug_close_eye_points, ibug_close_eye_trimesh

# label images
labeller(training_images, 'PTS', ibug_open_eye_points);
labeller(training_images, 'PTS', ibug_open_eye_trimesh);
training_images[0].landmarks['ibug_open_eye_trimesh'].lms.view_new()
```

```python
from pybug.aam.base import aam_builder
from pybug.transform.piecewiseaffine import PiecewiseAffineTransform

# set options
build_options = {'group': 'PTS',  
                 'trilist': training_images[0].landmarks['ibug_open_eye_trimesh'].lms.trilist,
                 'n_levels': 3,
                 'downscale': 2,
                 'features': ('igo', {'double_angles':True}),
                 'max_shape_components': 25,
                 'max_appearance_components': None}
# build aam
aam = aam_builder(training_images, **build_options)
```

```python
import numpy as np

aam.shape_models[0]._components = np.real(aam.shape_models[0]._components)
aam.shape_models[1]._components = np.real(aam.shape_models[1]._components)
aam.shape_models[2]._components = np.real(aam.shape_models[2]._components)
aam.shape_models[0]._eigenvalues = np.real(aam.shape_models[0]._eigenvalues)
aam.shape_models[1]._eigenvalues = np.real(aam.shape_models[1]._eigenvalues)
aam.shape_models[2]._eigenvalues = np.real(aam.shape_models[2]._eigenvalues)

# low resolution level
%matplotlib inline
aam.instance(level=0).view(channels=0) 
# middle resolution level
#aam.instance(level=1).view_new()
# highest resolution level
aam.instance().view_new(channels=0)
```

```python
import pybug.io as pio
from pybug.landmark import labeller, ibug_68_points, ibug_68_trimesh

test_images = []
# load landmarked imarges
for i in pio.import_images('/homes/ja310/Dropbox/Eye_Project/Exp1/EYES_AOM/New_images/open/Images/Right/*.png',):
    i.crop_to_landmarks_proportion(0.5)
    if i.n_channels == 3:
        test_images.append(i.as_greyscale(mode='luminosity'))
    else:
        test_images.append(i)
```

```python
from pybug.aam.fitter import LucasKanadeAAMFitter
from pybug.lucaskanade.appearance import AlternatingInverseCompositional
from pybug.lucaskanade.residual import LSIntensity
from pybug.transform.modeldriven import OrthoMDTransform
from pybug.transform.affine import SimilarityTransform

# set fitting options. As before we favour 
# clarity by explicitely stating all possible
# options
fitting_setting_options = \
    {'lk_algorithm': AlternatingInverseCompositional,
     'residual': LSIntensity,
     'md_transform_cls': OrthoMDTransform,
     'global_transform_cls': SimilarityTransform,
     'n_shape': [3, 6, 12], 
     'n_appearance': 25}

# initialize Lucas-Kanade aam fitting
lk_aam_fitter = LucasKanadeAAMFitter(aam, **fitting_setting_options)
```

```python
sm = aam.appearance_models[0]
np.sum(sm.eigenvalues_ratio[:100])
```

```python
fitting_options = {'group': 'PTS',
                   'label': 'all',
                   'runs': 1, 
                   'noise_std': 0.10,
                   'max_iters': 20,
                   'rotation': False,
                   'verbose': True, 
                   'view': True}

%matplotlib inline
# fit test images
fitting_list = lk_aam_fitter.fit_images(test_images[60:80], **fitting_options)
```

```python
%matplotlib inline
fitting_list.plot_error_dist(color_list=['r', 'b'], marker_list=['o', '*'])
```

```python
%matplotlib inline
fitting_list.plot_cumulative_error_dist(color_list=['r', 'b'], marker_list=['o', '*'])
```
