```python
import menpo.io as mio

images = []
for i in mio.import_images('/data/PhD/DataBases/cmu_car_data1_processed/data/view1/*', verbose=True):
    images.append(i)
```

```python
%matplotlib inline
images[7].view()
```

```python
from scipy.io import loadmat

ds = loadmat('/data/PhD/DataBases/cmu_car_data1_processed/FinalShape_view1_mine.mat')
landmarks = ds['Shape']
```

```python
from menpo.shape import PointCloud

grayscale_images = []
for j, i in enumerate(images):
    i.landmarks['PTS'] = PointCloud(landmarks[j][0]['landmarks'][0, 0].T[:, [1,0]])
    grayscale_images.append(i.as_greyscale())
```

```python
from menpo.visualize.widgets import browse_images

browse_images(images)
```

```python
from __future__ import division
import numpy as np
from menpo.fitmultilevel.aam import AAMBuilder
from menpo.fitmultilevel.featurefunctions import sparse_hog


limit = np.require(len(images)/2, dtype=np.int)
aam = AAMBuilder(feature_type=sparse_hog, normalization_diagonal=200, n_levels=1).build(grayscale_images[:limit], group='PTS', verbose=True)
```

```python
from menpo.visualize.widgets import visualize_aam

visualize_aam(aam)
print aam
```

```python
from menpo.fitmultilevel.aam import LucasKanadeAAMFitter

fitter = LucasKanadeAAMFitter(aam, n_shape=[3], n_appearance=0.5)
```

```python
print fitter
```

```python
fitting_results = []
for j, i in enumerate(grayscale_images[limit:limit+20]):
    gt_s = i.landmarks['PTS'].lms
    s = fitter.perturb_shape(gt_s, rotation=True, noise_std=0.08)
    print 'Image: ', j
    fr = fitter.fit(i, s, gt_shape=gt_s)
    fitting_results.append(fr)
```

```python
final_images = [fr.final_image for fr in fitting_results]
```

```python
from menpo.visualize.widgets import browse_final_images

browse_final_images(final_images)
```

```python
from menpo.visualize.widgets import plot_error_evolution_graph

plot_error_evolution_graph(fitting_results[5].errors())
```
