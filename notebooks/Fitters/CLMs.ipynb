```python
import menpo.io as pio
from menpo.landmark import labeller, ibug_68_trimesh, ibug_49_points

# load training images
training_images = []
for i in pio.import_images('/Users/joan/PhD/DataBases/lfpw/trainset/*.png'):
    # crop image
    i.crop_to_landmarks_proportion(0.2)
    # label it
    labeller(i, 'PTS', ibug_68_trimesh)
    # convert it to grayscale if needed
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
    # append it to the list
    training_images.append(i)
    
# load training images
for i in pio.import_images('/Users/joan/PhD/DataBases/helen/trainset/*.jpg', max_images=300):
    # crop image
    i.crop_to_landmarks_proportion(0.2)
    # label it
    labeller(i, 'PTS', ibug_68_trimesh)
    # convert it to grayscale if needed
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
    # append it to the list
    training_images.append(i)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.html.widgets import interact, interactive

def browse_images(images, group=None, label='all'):
    n = len(images)
    def view_image(i):
        images[i].landmarks[group][label].view()
        plt.show()
    interact(view_image, i=(0,n-1))
```

```python
browse_images(training_images, group='PTS')
```

```python
from menpo.fitmultilevel.clm import CLMBuilder
from menpo.fitmultilevel.featurefunctions import sparse_hog
from menpo.transform import PiecewiseAffine
from sklearn import qda

def random_forest(X, t):
    r"""
    """
    clf = a = qda.QDA()
    clf.fit(X, t)

    def random_forest_predict(x):
        return clf.predict_proba(x)[:, 1]

    return random_forest_predict

# options dictionary
options = {'classifier_type': random_forest,
           'feature_type': sparse_hog, 
           'diagonal_range': None,
           'patch_shape': (5, 5),
           'n_levels': 1,
           'scaled_levels': True,
           'max_shape_components': 25,
           'boundary': 3,
           'interpolator': 'scipy'
           }

# build aam
aam = CLMBuilder(**options).build(training_images, group='PTS')
```

```python
def visualize_clm(aam, bounds=(-3.0, 3.0)):    
    def view_aam(shape_pc1, shape_pc2, shape_pc3):
        aam.instance(shape_weights=[shape_pc1, shape_pc2, shape_pc3]).view(channels=0)
        plt.gcf().set_size_inches((5,5))
        plt.show()
    interact(view_aam, shape_pc1=bounds, shape_pc2=bounds, shape_pc3=bounds)
```

```python
plt.show()
def explore_fitting_result(fitting_result):    
    n = fitting_result.n_iters
    a = fitting_result.image.view()
    _ax = a.figure.axes[0]
    _line, = _ax.plot([], [], linestyle=' ')
    def view_fitting_result(it):
        shape = fitting_result.shapes(as_points=True)[it]
        _line.set_data(shape.points[:, 1], shape.points[:, 0])
        plt.show()
    interact(view_fitting_result, it=(0, n-1))
```
