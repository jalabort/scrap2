```python
import numpy as np

def linear_discriminant_decomposition(Xs, whiten=False, center=True, bias=False, inplace=False):
    C = len(Xs)
    
    mus = [np.mean(X, axis=0) for X in Xs]
    T = np.mean(np.asarray(mus), axis=0)
    
    Sigmas = []
    Omegas = []
    for (X, mu) in zip(Xs, mus):
        n_samples, n_features = X.shape
        
        if bias:
            N = n_samples
        else:
            N = n_samples - 1.0
        
        X = X - mu
        Sigmas.append(np.dot(X.T, X) / N)
        
        mu = mu - T
        Omegas.append(np.dot(mu[:, None], mu[None, :]) / C)
        
    S = 0
    for Sig in Sigmas:
        S += Sig
        
    O = 0
    for Ome in Omegas:
        O += Ome
        
    C = np.linalg.solve(S, O)
    
    from menpo.math.decomposition import eigenvalue_decomposition
    
    eigenvectors, eigenvalues = eigenvalue_decomposition(C)
        
    return eigenvectors, eigenvalues
```

```python
import numpy as np
from menpo.shape import PointCloud
import matplotlib.pylab as plt

plot_size = (6, 6)

patch_shape = (8, 8)
lim = patch_shape[0] / 2

centers = np.meshgrid(range(0, 100, 5), range(0, 100, 5)) 
centers = PointCloud(np.asarray([centers[0].flatten(), centers[1].flatten()]).T) 
```

```python
img = training_images[0].copy()
img.build_mask_around_landmarks(patch_shape)
img.view()
plt.gcf().set_size_inches(plot_size)
```

## 1st Level

```python
from menpo.fitmultilevel.functions import extract_local_patches_fast

patch_list = []
for i in training_images:
    patches = extract_local_patches_fast(i, i.landmarks['PTS'].lms, patch_shape)
    patches -= np.mean(patches, axis=(1, 2))[:, None, None, :]
    patch_list.append(patches)
        
patch_classes = []
for j in range(patches.shape[0]):
    patch_images = []
    for patches in patch_list:
        patch_img = patches[j, :, :, 0]
        patch_images.append(patch_img)
    patch_classes.append(np.reshape(np.asarray(patch_images), (-1, np.prod(patch_shape))))
```

```python
from menpo.model import PCAModel
from menpo.image import Image

eigenvectors, eigenvalues = linear_discriminant_decomposition(patch_classes)
patch_components = [Image(np.reshape(eigenvectors[:, j], patch_shape)) for j in range(8)]
```

```python
browse_images(patch_components)
```

```python
expanded_patch_pcs = []
for j in range(len(patch_components)):
    aux = np.zeros((img_shape[0]+patch_shape[0], img_shape[1]+patch_shape[1]))
    aux[-lim:, -lim:] = patch_components[j].pixels[:lim, :lim, 0]
    aux[:lim, -lim:] = patch_components[j].pixels[lim:, :lim, 0]
    aux[-lim:, :lim] = patch_components[j].pixels[:lim, lim:, 0]
    aux[:lim, :lim] = patch_components[j].pixels[lim:, lim:, 0]
    expanded_patch_pcs.append(Image(np.fft.fftshift(aux)))
```

```python
browse_images(expanded_patch_pcs)
```

```python
H1 = 0
fft_patch_pcs = []
for pc in expanded_patch_pcs:
    fft_pc = np.abs(np.fft.fft2(pc.pixels[..., 0]))
    fft_patch_pcs.append(Image(fft_pc[..., None]))
    H1 += fft_pc
H1 = Image(H1[..., None])
```

```python
browse_images(fft_patch_pcs)
```

```python
H1.view()
plt.gcf().set_size_inches(plot_size)
```

```python
expanded_images = []
for i in training_images:
    aux = np.zeros((img_shape[0]+patch_shape[0], img_shape[1]+patch_shape[1]))
    aux[-lim_y:, -lim_x:] = i.pixels[:lim_y, :lim_x, 0]
    aux[:lim_y, -lim_x:] = i.pixels[lim_y:, :lim_x, 0]
    aux[-lim_y:, :lim_x] = i.pixels[:lim_y, lim_x:, 0]
    aux[:lim_y, :lim_x] = i.pixels[lim_y:, lim_x:, 0]
    expanded_images.append(Image(np.fft.fftshift(aux)))
```

```python
browse_images(expanded_images)
```

```python
fft_images = []
filtered_fft_images = []
final_response_images = []
for (exp_i, i) in zip(expanded_images, training_images):
    fft_img = Image(np.fft.fft2(exp_i.pixels[..., 0]))
    filtered_fft_img = Image(fft_img.pixels[..., 0] * H1.pixels[..., 0]**2)
    response_img = Image(np.real(np.fft.ifft2(filtered_fft_img.pixels[..., 0])[patch_shape[0]/2:-patch_shape[0]/2, patch_shape[1]/2:-patch_shape[1]/2]))
    response_img.landmarks = i.landmarks
    fft_images.append(fft_img)
    filtered_fft_images.append(filtered_fft_img)
    final_response_images.append(response_img)
```

```python
response_images = []
for (i, fft_i) in zip(training_images, fft_images):
    for fft_pc in fft_patch_pcs:
        filtered_fft_img = Image(fft_i.pixels[..., 0] * fft_pc.pixels[..., 0])
        response_img = Image(np.real(np.fft.ifft2(filtered_fft_img.pixels[..., 0])[patch_shape[0]/2:-patch_shape[0]/2, patch_shape[1]/2:-patch_shape[1]/2]))
        response_img.landmarks = i.landmarks
        response_images.append(response_img)
```
