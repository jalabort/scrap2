## 1st Level

```python
class DeepConvKernel

    def __init__(filter_builder, n_levels=10, n_filters=8)
        self.filter_builder = filter_builder
        self.n_levels = n_levels
        self.n_filters = n_filters

    def build(images, centers)
        for level in range(self.n_levels)
            patches = extract_patches(images, centers)
            filters = self.filter_builder(patches, n_filters)
    
    def extract_patches(images, centers)
        patches = []
        for (i, c) in zip(images, centers):
            ps = extract_local_patches_fast(i, c, self.patch_shape)
            if self.mean_center:
                ps -= np.mean(ps, axis=(1, 2))[:, None, None, :]
            for p in ps:
                p = Image(p[:, :, 0])
                patches.append(p)
        return patches
```

```python
from __future__ import division

def extract_local_patches_fast(image, centres, patch_shape, out=None,
                               dtype=None):
    r"""
    Extract square patches from an image about centres.

    Parameters
    ----------
    image : :map:`Image`
        The image to extract patches from

    centres : :map:`PointCloud`
        The centres around which the patches should be extracted

    patch_shape : `tuple` of `ints`
        The size of the patch in each dimension

    out : `ndarray` of shape: ``n_centres, + patch_shape + n_channels,``,
    optional
        The output array to be assigned to. If `None`, a new numpy array
        will be created.

    Returns
    -------
    patches : `ndarray` of shape: ``n_centres, + patch_shape + n_channels,``
        The patches as a single numpy array.
    """
    if out is not None:
        patches = out
    else:
        patches = np.empty(
            (centres.n_points,) + patch_shape + (image.n_channels,),
            dtype=dtype)
    # 0 out the patches array
    patches[...] = 0
    image_size = np.array(image.shape, dtype=np.int)
    patch_shape = np.array(patch_shape, dtype=np.int)
    centres = np.require(centres.points, dtype=np.int)
    half_patch_shape = np.require(np.floor(patch_shape / 2), dtype=np.int)
    add_to_patch = np.mod(patch_shape, 2) 
    # 1. compute the extents
    c_min = centres - half_patch_shape
    c_max = centres + half_patch_shape + add_to_patch
    out_min_min = c_min < 0
    out_min_max = c_min > image_size
    out_max_min = c_max < 0
    out_max_max = c_max > image_size

    # 1. Build the extraction slices
    ext_s_min = c_min.copy()
    ext_s_max = c_max.copy()
    # Clamp the min to 0
    ext_s_min[out_min_min] = 0
    ext_s_max[out_max_min] = 0
    # Clamp the max to image bounds across each dimension
    for i in xrange(image.n_dims):
        ext_s_max[out_max_max[:, i], i] = image_size[i] - 1
        ext_s_min[out_min_max[:, i], i] = image_size[i] - 1

    # 2. Build the insertion slices
    ins_s_min = ext_s_min - c_min
    ins_s_max = np.maximum(ext_s_max - c_max + patch_shape, (0, 0))

    for i, (e_a, e_b, i_a, i_b) in enumerate(zip(ext_s_min, ext_s_max,
                                                 ins_s_min, ins_s_max)):
        # build a list of insertion slices and extraction slices
        i_slices = [slice(a, b) for a, b in zip(i_a, i_b)]
        e_slices = [slice(a, b) for a, b in zip(e_a, e_b)]
        # get a view onto the patch we are on
        patch = patches[i, ...]
        # apply the slices to map
        patch[i_slices] = image.pixels[e_slices]

    return patches
```

```python
from menpo.image import Image
from menpo.fitmultilevel.functions import extract_local_patches_fast

patch_images = []
for i in training_images:
    patches = extract_local_patches_fast(i, i.landmarks['PTS'].lms, patch_shape)
    patches -= np.mean(patches, axis=(1, 2))[:, None, None, :]
    for patch in patches:
        patch_img = Image(patch[:, :, 0])
        patch_images.append(patch_img)
```

```python
from menpo.model import PCAModel

patch_pca = PCAModel(patch_images, center=False)
patch_pca.n_active_components = 0.98
print patch_pca
```

```python
patch_pcs= [patch_pca.component(j) for j in range(patch_pca.n_active_components)]
```

```python
browse_images(patch_pcs)
```

```python
expanded_patch_pcs = []
for j in np.arange(patch_pca.n_active_components):
    aux = np.zeros((img_shape[0]+patch_shape[0], img_shape[1]+patch_shape[1]))
    aux[-lim:, -lim:] = patch_pca.component(j).pixels[:lim, :lim, 0]
    aux[:lim, -lim:] = patch_pca.component(j).pixels[lim:, :lim, 0]
    aux[-lim:, :lim] = patch_pca.component(j).pixels[:lim, lim:, 0]
    aux[:lim, :lim] = patch_pca.component(j).pixels[lim:, lim:, 0]
    expanded_patch_pcs.append(Image(np.fft.fftshift(aux)))
```

```python
browse_images(expanded_patch_pcs)
```

```python
H1 = 0
fft_patch_pcs = []
for pc in expanded_patch_pcs:
    fft_pc = np.fft.fft2(pc.pixels[..., 0])
    fft_patch_pcs.append(Image(fft_pc[..., None]))
    H1 += np.abs(fft_pc)
H1 = Image(H1[..., None])
```

```python
browse_images(fft_patch_pcs)
```

```python
H1.view()
plt.gcf().set_size_inches(plot_size)
```

```python
expanded_images = []
for i in training_images:
    aux = np.zeros((img_shape[0]+patch_shape[0], img_shape[1]+patch_shape[1]))
    aux[-lim_y:, -lim_x:] = i.pixels[:lim_y, :lim_x, 0]
    aux[:lim_y, -lim_x:] = i.pixels[lim_y:, :lim_x, 0]
    aux[-lim_y:, :lim_x] = i.pixels[:lim_y, lim_x:, 0]
    aux[:lim_y, :lim_x] = i.pixels[lim_y:, lim_x:, 0]
    expanded_images.append(Image(np.fft.fftshift(aux)))
```

```python
browse_images(expanded_images)
```

```python
fft_images = []
filtered_fft_images = []
final_response_images = []
for (exp_i, i) in zip(expanded_images, training_images):
    fft_img = Image(np.fft.fft2(exp_i.pixels[..., 0]))
    filtered_fft_img = Image(fft_img.pixels[..., 0] * H1.pixels[..., 0]**0.5)
    response_img = Image(np.real(np.fft.ifft2(filtered_fft_img.pixels[..., 0])[patch_shape[0]/2:-patch_shape[0]/2, patch_shape[1]/2:-patch_shape[1]/2]))
    response_img.landmarks = i.landmarks
    fft_images.append(fft_img)
    filtered_fft_images.append(filtered_fft_img)
    final_response_images.append(response_img)
```
