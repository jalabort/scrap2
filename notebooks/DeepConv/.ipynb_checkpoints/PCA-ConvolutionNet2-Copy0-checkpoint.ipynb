```python
import menpo.io as mio
import matplotlib.pylab as plt
plot_size = (6, 6)

# load training images
training_images = []
for i in mio.import_images('/data/PhD/DataBases/lfpw/trainset/*', verbose=True):
    # crop image
    i.crop_to_landmarks_proportion_inplace(0.2)
    # convert it to grayscale if needed
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
    # append it to the list
    training_images.append(i)
    
for i in mio.import_images('/data/PhD/DataBases/lfpw/testset/*', verbose=True):
    # crop image
    i.crop_to_landmarks_proportion_inplace(0.2)
    # convert it to grayscale if needed
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
    # append it to the list
    training_images.append(i)
    
for i in mio.import_images('/data/PhD/DataBases/ibug/*', verbose=True):
    # crop image
    i.crop_to_landmarks_proportion_inplace(0.2)
    # convert it to grayscale if needed
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
    # append it to the list
    training_images.append(i)
    
for i in mio.import_images('/data/PhD/DataBases/afw/*', verbose=True):
    # crop image
    i.crop_to_landmarks_proportion_inplace(0.2)
    # convert it to grayscale if needed
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
    # append it to the list
    training_images.append(i)
```

```python
from menpo.visualize.widgets import browse_images

%matplotlib inline
browse_images(training_images)
```

```python
diagonal_range = 200
training_images = [i.rescale_landmarks_to_diagonal_range(diagonal_range) for i in training_images] 
```

```python
%matplotlib inline
browse_images(training_images)
```

```python
import numpy as np
from menpo.shape import PointCloud

patch_shape = np.array((8, 8))
lim = patch_shape[0] / 2

patch_shape2 = np.array((4, 4))
lim2 = patch_shape2[0] / 2

extended_patch_shape = 5*patch_shape
extended_lim = extended_patch_shape[0] / 2

centers = np.meshgrid(range(0, 100, 5), range(0, 100, 5)) 
centers = PointCloud(np.asarray([centers[0].flatten(), centers[1].flatten()]).T) 
```

```python
img = training_images[0].copy()
img.build_mask_around_landmarks(patch_shape)
img.view()
plt.gcf().set_size_inches(plot_size)
```

## 1st Level

```python
from menpo.image import Image
from menpo.fitmultilevel.functions import extract_local_patches_fast

extended_patch_images = []
patch_images = []
centered_patch_images = []
for i in training_images:
    extended_patches = extract_local_patches_fast(i, i.landmarks['PTS'].lms, tuple(extended_patch_shape))
    for extended_patch in extended_patches:
        extended_patch_img = Image(extended_patch[:, :, 0])
        extended_patch_images.append(extended_patch_img)
        patch_img = Image(extended_patch[(extended_patch_shape[0] - patch_shape[0])/2:-(extended_patch_shape[0] - patch_shape[0])/2, (extended_patch_shape[1] - patch_shape[1])/2:-(extended_patch_shape[1] - patch_shape[1])/2, 0])
        patch_images.append(patch_img)
        
        patch_img2 = patch_img.copy()
        patch_img2.pixels[..., 0] -= np.mean(patch_img2.pixels[..., 0])[None, None]
        centered_patch_images.append(patch_img2)

extended_patch_images2 = []
patch_images2 = []
centered_patch_images2 = []
for i in training_images:
    extended_patches = extract_local_patches_fast(i, i.landmarks['PTS'].lms, tuple(extended_patch_shape))
    for extended_patch in extended_patches:
        extended_patch_img = Image(extended_patch[:, :, 0])
        extended_patch_images2.append(extended_patch_img)
        patch_img = Image(extended_patch[(extended_patch_shape[0] - patch_shape2[0])/2:-(extended_patch_shape[0] - patch_shape2[0])/2, (extended_patch_shape[1] - patch_shape2[1])/2:-(extended_patch_shape[1] - patch_shape2[1])/2, 0])
        patch_images2.append(patch_img)
        
        patch_img2 = patch_img.copy()
        patch_img2.pixels[..., 0] -= np.mean(patch_img2.pixels[..., 0])[None, None]
        centered_patch_images2.append(patch_img2)
```

```python
expanded_patch_pcs = []
for j in np.arange(patch_pca.n_active_components):
    aux = np.zeros((extended_patch_shape[0], extended_patch_shape[1]))
    aux[-lim:, -lim:] = patch_pca.component(j).pixels[:lim, :lim, 0]
    aux[:lim, -lim:] = patch_pca.component(j).pixels[lim:, :lim, 0]
    aux[-lim:, :lim] = patch_pca.component(j).pixels[:lim, lim:, 0]
    aux[:lim, :lim] = patch_pca.component(j).pixels[lim:, lim:, 0]
    expanded_patch_pcs.append(Image(np.fft.fftshift(aux)))
    
for j in np.arange(patch_pca2.n_active_components):
    aux = np.zeros((extended_patch_shape[0], extended_patch_shape[1]))
    aux[-lim2:, -lim2:] = patch_pca2.component(j).pixels[:lim2, :lim2, 0]
    aux[:lim2, -lim2:] = patch_pca2.component(j).pixels[lim2:, :lim2, 0]
    aux[-lim2:, :lim2] = patch_pca2.component(j).pixels[:lim2, lim2:, 0]
    aux[:lim2, :lim2] = patch_pca2.component(j).pixels[lim2:, lim2:, 0]
    expanded_patch_pcs.append(Image(np.fft.fftshift(aux)))
    
expanded_patch_pcs_list = [expanded_patch_pcs]
```

```python
H1 = 0
fft_patch_pcs = []
for pc in expanded_patch_pcs:
    fft_pc = np.abs(np.fft.fft2(pc.pixels[..., 0]))
    fft_patch_pcs.append(Image(fft_pc[..., None]))
    H1 += fft_pc
H1 = Image(H1[..., None])
```

```python
fft_patches = []
#filtered_fft_patches = []
extended_final_response_patches = []
centered_final_response_patches = []
centered_final_response_patches2 = []
final_response_patches = []
for ext_i in extended_patch_images:
    # fft
    fft_patch = Image(np.fft.fft2(ext_i.pixels[..., 0]))
    fft_patches.append(fft_patch)
    # filtered
    filtered_fft_patch = Image(fft_patch.pixels[..., 0] * H1.pixels[..., 0] ** 0.5)
    #filtered_fft_images.append(filtered_fft_img)
    # response
    extended_response_patch = Image(np.real(np.fft.ifft2(filtered_fft_patch.pixels[..., 0]))) 
    extended_final_response_patches.append(extended_response_patch)
    
    response_patch = Image(extended_response_patch.pixels[(extended_patch_shape[0] - patch_shape[0])/2:-(extended_patch_shape[0] - patch_shape[0])/2, (extended_patch_shape[1] - patch_shape[1])/2:-(extended_patch_shape[1] - patch_shape[1])/2, 0])
    final_response_patches.append(response_patch)
    
    response_patch2 = response_patch.copy()
    response_patch2.pixels[..., 0] -= np.mean(response_patch2.pixels[..., 0])[None, None]
    centered_final_response_patches.append(response_patch2)
    
    response_patch = Image(extended_response_patch.pixels[(extended_patch_shape[0] - patch_shape2[0])/2:-(extended_patch_shape[0] - patch_shape2[0])/2, (extended_patch_shape[1] - patch_shape2[1])/2:-(extended_patch_shape[1] - patch_shape2[1])/2, 0])
    
    response_patch3 = response_patch.copy()
    response_patch3.pixels[..., 0] -= np.mean(response_patch3.pixels[..., 0])[None, None]
    centered_final_response_patches2.append(response_patch3)
    
```

## Alignment

```python
img_shape = (170, 150)

Hs = []
for expanded_patch_pcs in expanded_patch_pcs_list:
    H = 0
    for patch_pc in expanded_patch_pcs:
        aux = np.zeros(img_shape)
        aux[-extended_lim:, -extended_lim:] = patch_pc.pixels[:extended_lim, :extended_lim, 0]
        aux[:extended_lim, -extended_lim:] = patch_pc.pixels[extended_lim:, :extended_lim, 0]
        aux[-extended_lim:, :extended_lim] = patch_pc.pixels[:extended_lim, extended_lim:, 0]
        aux[:extended_lim, :extended_lim] = patch_pc.pixels[extended_lim:, extended_lim:, 0]
        fft_pc = np.abs(np.fft.fft2(np.fft.fftshift(aux)))
        H += fft_pc
    Hs.append(H)
    
H = Hs[0]
for h in Hs[1:]:
    H *= h
```
