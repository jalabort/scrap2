```python
import pybug.io as pio
from pybug.landmark import labeller, ibug_66_points

training_images = []
# load landmarked imarges
for i in pio.import_images('/Users/joan/PhD/DataBases/lfpw/trainset/*.png',
                     max_images=811):
    i.crop_to_landmarks_proportion(0.4)
    if i.n_channels == 3:
        training_images.append(i.as_greyscale(mode='luminosity'))
    else:
        training_images.append(i)
```

```python
from pybug.sdm.base import sdm_builder

def hog_closure(image):
    return image.features.hog(mode='sparse', cell_size=8, block_size=2)

# build aam
sdm = sdm_builder(training_images, group='ibug_66_points', patch_shape=(15, 15), diagonal_range=100, n_levels=2,
                  downscale=1.1, scaled_levels=True, feature_type=hog_closure, 
                  noise_std=0.04, rotation=False, n_perturbations=20)
```
