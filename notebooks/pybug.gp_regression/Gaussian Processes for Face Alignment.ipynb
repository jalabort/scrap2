# Gaussian Processes for Face Alignment

```python
# import
import numpy as np
import pybug.io as pio
from pybug.landmark import labeller, ibug_68_trimesh
from pybug.aam.functions import mean_pointcloud
```

```python
# load the fpw database
training_images = []
for i in pio.import_images('/data/PhD/DataBases/lfpw/trainset/*.png', max_images=811):
    i.crop_to_landmarks_proportion(0.3)
    if i.n_channels == 3:
        training_images.append(i.as_greyscale(mode='luminosity'))
    else:
        training_images.append(i)
```

```python
%matplotlib inline

#visualize the first image
training_images[0].landmarks['PTS'].view_new()
```

```python
# extract the original landmarks
training_landmarks = [i.landmarks['PTS'].lms for i in training_images]
# compute the reference landmarks as the mean of the previous landmarks
reference_landmarks = mean_pointcloud(training_landmarks)
```

```python
# rescale the training images
training_images = [i.rescale_to_reference_landmarks(reference_landmarks, group='PTS') 
                   for i in training_images]
```

```python
# set downscale
downscale = 2

# set pyramid generators
pyramid_generator = [i.gaussian_pyramid(downscale=downscale) for i in training_images]
# obtain and reverse the pyramid of training images
training_image_pyramid = [[i.next() for i in pyramid_generator] for _ in np.arange(0,3)]
training_image_pyramid.reverse()

# extract the pyramid of training shapes
training_landmarks_pyramid = [[i.landmarks['PTS'].lms for i in p] for p in training_image_pyramid]
```

```python
# set hog options
features_dic = {'type':'hog', 
                'options': {'num_bins':9,
                            'cell_size':4, 
                            'block_size':2,
                            'window_step_vertical': 4, 
                            'window_step_horizontal': 4}}
```

```python
# set daisy options
features_dic = {'type':'daisy', 
                'options': {'step': 4,
                            'radius': 4, 
                            'rings': 4, 
                            'histograms': 8, 
                            'orientations': 8}}
```

```python
# set lbp options
features_dic = {'type':'lbp'}
```

```python
# set igo options
features_dic = {'type':'igo',
                'options': {'double_angles': False}}
```

```python
# set norm options
features_dic = {'type':'norm',
                'options': {'mode': 'all'}}
```

```python
# set raw options
features_dic = {'type':'raw'}
```

```python
# other options
patch_size = ((8, 8), (16,16), (16,16))
noise_std = 0.02
n_perturbations = 5
```

```python
%matplotlib inline

# visualize the first image of the first pyramid
training_image_pyramid[0][0].build_mask_around_landmarks(patch_size[0])
training_image_pyramid[0][0].landmarks['PTS'].view_new()

#visualize the first image of the first pyramid
training_image_pyramid[1][0].build_mask_around_landmarks(patch_size[1])
training_image_pyramid[1][0].landmarks['PTS'].view_new()

#visualize the first image of the first pyramid
training_image_pyramid[2][0].build_mask_around_landmarks(patch_size[2])
training_image_pyramid[2][0].landmarks['PTS'].view_new()
```

```python
from pybug.gp_regression.base import NonParametricRegressor

# build nn parametric regressor
npr0 = NonParametricRegressor(reference_landmarks, 
                              features_dic=features_dic, 
                              patch_size=patch_size[0],
                              noise_std=noise_std,
                              n_perturbations=n_perturbations)
```

```python
# train it
npr0.train_gp(training_image_pyramid[0], training_landmarks_pyramid[0], kernel_type='linear')
```

```python
test_images = []
# load landmarked imarges
for i in pio.import_images('/data/PhD/DataBases/afw/*.jpg', max_images=224):
    i.crop_to_landmarks_proportion(0.3)
    if i.n_channels == 3:
        test_images.append(i.as_greyscale(mode='luminosity'))
    else:
        test_images.append(i)
```

```python
%matplotlib inline

#visualize the first image
test_images[0].landmarks['PTS'].view_new()
```

```python
# rescale the training images
test_images = [i.rescale_to_reference_landmarks(reference_landmarks, group='PTS') 
               for i in test_images]
```

```python
# set pyramid generators
pyramid_generator = [i.gaussian_pyramid(downscale=downscale) for i in test_images]
# obtain and reverse the pyramid of test images
test_image_pyramid = [[i.next() for i in pyramid_generator] for _ in np.arange(0,3)]
test_image_pyramid.reverse()

# extract the pyramid of test shapes
test_landmarks_pyramid = [[i.landmarks['PTS'].lms for i in p] for p in test_image_pyramid]

# perturb the original landmarks
perturbed_landmarks_pyramid = [[npr0._perturb_landmarks(s) for s in p] for p in test_landmarks_pyramid]
```

```python
ind = 6
fitted_shape = npr0.align_gp(test_image_pyramid[0][ind], perturbed_landmarks_pyramid[0][ind])
```

```python
test_image_pyramid[0][ind].landmarks['fitted'] = fitted_shape
test_image_pyramid[0][ind].landmarks['initial'] = perturbed_landmarks_pyramid[0][ind]
test_image_pyramid[0][ind].landmarks['gt'] = test_landmarks_pyramid[0][ind]
test_image_pyramid[0][ind].landmarks['initial'].view() 
test_image_pyramid[0][ind].landmarks['fitted'].view_new()
```
