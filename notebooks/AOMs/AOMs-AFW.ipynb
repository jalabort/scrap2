```python
%matplotlib inline

import menpo.io as mio
from menpo.landmark import labeller, ibug_68_trimesh, ibug_68_points

# load training images
training_images = []
for i in mio.import_images('/Users/joan/PhD/DataBases/lfpw/*/*', verbose=True):
    # crop image
    i.crop_to_landmarks_proportion_inplace(1)
    # label it
    labeller(i, 'PTS', ibug_68_trimesh)
    labeller(i, 'PTS', ibug_68_points)
    # convert it to grayscale if needed
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
    # append it to the list
    training_images.append(i)
```

```python
from menpo.fitmultilevel.aam import AAMBuilder

def double_igo(img):
    return img.features.igo(double_angles=True)

# build aam
aambuilder = AAMBuilder(feature_type=double_igo, normalization_diagonal=120, downscale=1.5, 
                        trilist=training_images[0].landmarks['ibug_68_trimesh'].lms.trilist)
aam = aambuilder.build(training_images, group='PTS', verbose=True)
```

```python
from menpo.fitmultilevel.aam import LucasKanadeAAMFitter

fitter = LucasKanadeAAMFitter(aam, n_shape=[3, 6, 12], n_appearance=[50, 100, 150])
```

```python
import menpo.io as mio

# load test images
test_images = []
for i in mio.import_images('/Users/joan/PhD/DataBases/afw/*', verbose=True):
    # crop image
    i.crop_to_landmarks_proportion_inplace(0.5)
    # convert it to grayscale if needed
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
    # append it to the list
    test_images.append(i)
```

```python
fitting_results = []
for j, i in enumerate(test_images):
    gt_s = i.landmarks['PTS'].lms
    s = fitter.perturb_shape(gt_s, noise_std=0.04)
    fr = fitter.fit(i, s, gt_shape=gt_s)
    fitting_results.append(fr)
    print 'Image: ', j
```

```python
import matplotlib.pylab as plt
from IPython.html.widgets import interact

import numpy as np

def browse_final_images2(test_images, final_shapes, plot_size=(12, 12)):
    n_images = len(test_images)

    def view_final_image(ind):
        shape = final_shapes[ind].points
        image = test_images[ind]
        
        image.view()
        
        points = shape[0:17, :]
        plt.plot(points[:, 1], points[:, 0], linewidth=2, marker='s', color='b')
        
        points = shape[17:22, :]
        plt.plot(points[:, 1], points[:, 0], linewidth=2, marker='s', color='b')
        
        points = shape[22:27, :]
        plt.plot(points[:, 1], points[:, 0], linewidth=2, marker='s', color='b')
        
        points = shape[27:31, :]
        plt.plot(points[:, 1], points[:, 0], linewidth=2, marker='s', color='b')
        
        points = shape[31:36, :]
        plt.plot(points[:, 1], points[:, 0], linewidth=2, marker='s', color='b')
        
        points = np.vstack((shape[36:42, :], shape[36, :]))
        plt.plot(points[:, 1], points[:, 0], linewidth=2, marker='s', color='b')
    
        points = np.vstack((shape[42:48, :], shape[42, :]))
        plt.plot(points[:, 1], points[:, 0], linewidth=2, marker='s', color='b')
        
        points = np.vstack((shape[48:60, :], shape[48, :]))
        plt.plot(points[:, 1], points[:, 0], linewidth=2, marker='s', color='b')
        
        points = np.vstack((shape[60:68, :], shape[60, :]))
        plt.plot(points[:, 1], points[:, 0], linewidth=2, marker='s', color='b')
        
        plt.gcf().set_size_inches(plot_size)
        plt.axis('off')

    interact(view_final_image, ind=(0, n_images-1))
```
