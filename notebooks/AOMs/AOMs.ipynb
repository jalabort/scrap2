```python
%matplotlib inline

import menpo.io as mio
from menpo.landmark import labeller, ibug_68_trimesh, ibug_49_points

# load training images
training_images = []
for i in mio.import_images('/Users/joan/PhD/DataBases/lfpw/trainset/*', verbose=True):
    # crop image
    i.crop_to_landmarks_proportion_inplace(1)
    # label it
    labeller(i, 'PTS', ibug_68_trimesh)
    labeller(i, 'PTS', ibug_49_points)
    # convert it to grayscale if needed
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
    # append it to the list
    training_images.append(i)
```

```python
from menpo.fitmultilevel.aam import AAMBuilder

def double_igo(img):
    return img.features.igo(double_angles=True)

# build aam
aambuilder = AAMBuilder(feature_type='igo', normalization_diagonal=100, downscale=1.5, 
                        trilist=training_images[0].landmarks['ibug_68_trimesh'].lms.trilist)
aam = aambuilder.build(training_images, group='PTS', verbose=True)
```

```python
from menpo.fitmultilevel.aam import LucasKanadeAAMFitter

fitter = LucasKanadeAAMFitter(aam, n_shape=[3, 6, 12], n_appearance=100)
```

```python
import menpo.io as mio

# load test images
test_images = []
for i in mio.import_images('/Users/joan/PhD/DataBases/afw/*', verbose=True):
    # crop image
    i.crop_to_landmarks_proportion_inplace(0.5)
    # convert it to grayscale if needed
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
    # append it to the list
    test_images.append(i)
```

```python
fitting_results = []
for j, i in enumerate(test_images):
    gt_s = i.landmarks['PTS'].lms
    s = fitter.perturb_shape(gt_s, noise_std=0.02)
    
    print 'Image: ', j
    fr = fitter.fit(i, s, gt_shape=gt_s)
    fitting_results.append(fr)
```

```python
final_images = [fr.final_image for fr in fitting_results]
```

```python
from menpo.visualize.widgets import browse_final_images

browse_final_images(final_images)
```
