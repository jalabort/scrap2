## Load training data

```python
import menpo.io as mio
from menpo.landmark import labeller, ibug_face_66
from cvpr2015.utils import convert_from_menpo

training_images = []
for i in mio.import_images('/data/PhD/DataBases/faces/lfpw/trainset/', verbose=True, max_images=200):
    
    # convert the image from menpo Image to cvpr2015 Image (channels at front)
    i = convert_from_menpo(i)
    
    labeller(i, 'PTS', ibug_face_66)
    i.crop_to_landmarks_proportion_inplace(0.5, group='PTS')
    if i.n_channels == 3:
        i = i.as_greyscale(mode='average')
    training_images.append(i)
```

```python
from menpo.visualize import visualize_images

visualize_images(training_images)
```

## Load test data

```python
import menpo.io as mio
from menpo.landmark import labeller, ibug_face_66
from cvpr2015.utils import convert_from_menpo

test_images = []
for i in mio.import_images('/data/PhD/DataBases/faces/lfpw/testset/', verbose=True, max_images=None):
    
    # convert the image from menpo Image to fg2015 Image (channels at front)
    i = convert_from_menpo(i)
    
    i.rescale_landmarks_to_diagonal_range(200)
    i.crop_to_landmarks_proportion_inplace(0.5)
    labeller(i, 'PTS', ibug_face_66)
    if i.n_channels == 3:
        i = i.as_greyscale(mode='average')
    test_images.append(i)
```

```python
from menpo.visualize import visualize_images

visualize_images(test_images)
```

## Active Appearance Models

#### Build

```python
from cvpr2015.feature import no_op, aam_dsift
from cvpr2015.deformablemodel import PartsAAMBuilder

builder = PartsAAMBuilder(parts_shape=(17, 17), features=aam_dsift, diagonal=100, 
                          scale_shapes=False, scales=(1, .5))
aam = builder.build(training_images, group='ibug_face_66', verbose=True)
```

#### Test

```python
from cvpr2015.deformablemodel import PartsAAMFitter
from cvpr2015.deformablemodel.aam import AIC, PIC

sampling_mask = np.require(np.zeros((17, 17)), dtype=np.bool)
sampling_mask[1::4, 1::4] = True
                  
fitter = PartsAAMFitter(aam, algorithm_cls=AIC, n_shape=[3, 12], 
                        n_appearance=[25, 50], sampling_mask=sampling_mask)
```

```python
np.random.seed(seed=1)

fitter_results = []

for j, i in enumerate(test_images[:]):
    
    gt_s = i.landmarks['ibug_face_66'].lms
    s = fitter.perturb_shape(gt_s, noise_std=0.0)
    
    fr = fitter.fit(i, s, gt_shape=gt_s, max_iters=20)
    fr.downscale = 0.5
    
    fitter_results.append(fr)
    
    print 'Image: ', j
    print fr
```

#### Visualize Results

```python
from menpofit.visualize import visualize_fitting_results
    
visualize_fitting_results(fitter_results)
```
