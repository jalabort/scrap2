## Load training data

```python
import menpo.io as mio
from menpo.landmark import labeller, ibug_face_68, ibug_face_68_trimesh

training_images = []
for i in mio.import_images('/data/PhD/DataBases/faces/lfpw/trainset/', 
                           max_images=10, verbose=True):
    labeller(i, 'PTS', ibug_face_68)
    labeller(i, 'PTS', ibug_face_68_trimesh)
    i.crop_to_landmarks_proportion_inplace(0.5, group='ibug_face_68')
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
    training_images.append(i)
```

```python
from menpo.visualize import browse_images

browse_images(training_images)
```

## Build Menpo-AAM

```python
from menpo.fitmultilevel.aam import AAMBuilder
from menpo.transform import ThinPlateSplines

def norm_int(image):
    norm_image = image.copy()
    norm_image.normalize_norm_inplace()
    return norm_image

aam_builder = AAMBuilder(
    normalization_diagonal=100, feature_type=None, n_levels=2,
    trilist=i.landmarks['ibug_face_68_trimesh'].lms.trilist, downscale=2)

aam = aam_builder.build(
    training_images, group='ibug_face_68', verbose=True)
```

```python
print aam
```

```python
from menpo.visualize import visualize_aam

visualize_aam(aam)
```

## Build Paper-AAM

```python
from menpo.advanced_aam_paper import GlobalAAMBuilder

gaam_builder = GlobalAAMBuilder(
    feature_type=None, diagonal=100, scales=[1, .5],
    trilist=i.landmarks['ibug_face_68_trimesh'].lms.trilist)
                                      
gaam = gaam_builder.build(
    training_images, group='ibug_face_68', verbose=True)
```
