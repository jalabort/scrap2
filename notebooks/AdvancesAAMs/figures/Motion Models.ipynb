```python
import menpo.io as mio
from menpo.landmark import labeller, ibug_face_68_trimesh
from menpo.aam.image import MaskedImage

images = []
for i in mio.import_images('/data/PhD/DataBases/faces/ibug/', 
                           max_images=10, verbose=True):
    
    new_i = MaskedImage(np.rollaxis(i.pixels, -1))
    new_i.landmarks = i.landmarks
    
    labeller(new_i, 'PTS', ibug_face_68_trimesh)
    new_i.crop_to_landmarks_proportion_inplace(0.5, group='PTS')
    if new_i.n_channels == 3:
        new_i = new_i.as_greyscale(mode='average')
    images.append(new_i)
```

```python
import dill as pickle

aam = pickle.load(open('/data/PhD/Models/pwa_aam.menpo', 'r'))
```

```python
am = aam.appearance_models[-1]
```

```python
from menpo.aam.image import BooleanImage
from menpo.landmark import labeller, ibug_face_68_trimesh

rf = BooleanImage.blank(am.mean.shape)
rf.landmarks['source'] = am.mean.landmarks['source']
labeller(rf, 'source', ibug_face_68_trimesh)

s0 = rf.landmarks['ibug_face_68_trimesh'].lms
points = s0.points
tri = s0.trilist
```

```python
scatter(rf.true_indices[:, 1], -rf.true_indices[:, 0], marker= '.', color='black', s=5)
axis('off')
triplot(points[:, 1], -points[:, 0], tri, color='b', linewidth=2)
triplot(points[:, 1], -points[:, 0], tri[10,:][None, ...], color='red', linewidth=4)

rf.landmarks['ibug_face_68_trimesh']

gcf().set_size_inches(16,16)
```

```python
from menpo.transform import PiecewiseAffine

s = images[8].landmarks['ibug_face_68_trimesh'].lms

t = PiecewiseAffine(s0, s)
```

```python
true_indices = t.apply(am.mean.mask.true_indices)
```

```python
scatter(true_indices[:, 1], -true_indices[:, 0], marker= '.', color='black', s=2)
axis('off')
triplot(s.points[:, 1], -s.points[:, 0], tri, color='b', linewidth=2)
triplot(s.points[:, 1], -s.points[:, 0], tri[10,:][None, ...], color='red', linewidth=4)

gcf().set_size_inches(16,16)
```

```python
images[8].view()
scatter(true_indices[:, 1], true_indices[:, 0], marker= '.', color='black', s=2)
axis('off')
triplot(s.points[:, 1], s.points[:, 0], tri, color='b', linewidth=2)
triplot(s.points[:, 1], s.points[:, 0], tri[10,:][None, ...], color='red', linewidth=4)

gcf().set_size_inches(16,16)
```

```python
images[8].warp_to(am.mean.mask, t).view()
scatter(rf.true_indices[:, 1], rf.true_indices[:, 0], marker= '.', color='black', s=5)
axis('off')
triplot(points[:, 1], points[:, 0], tri, color='b', linewidth=2)
triplot(points[:, 1], points[:, 0], tri[10,:][None, ...], color='red', linewidth=4)

rf.landmarks['ibug_face_68_trimesh']

gcf().set_size_inches(16,16)
```
