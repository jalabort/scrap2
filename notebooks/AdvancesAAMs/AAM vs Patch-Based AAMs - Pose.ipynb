## Load training data

```python
import menpo.io as mio
from menpo.landmark import labeller, stickmen_pose

training_images = []
for i in mio.import_images('/data/PhD/DataBases/human-pose/buffy_stickmen_v3.01/buffy_s5e2_original/', 
                           max_images=None, verbose=True):
    labeller(i, 'LJSON', stickmen_pose)
    i.crop_to_landmarks_proportion_inplace(
        0.5, group='stickmen_pose')
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
        training_images.append(i)
```

```python
from menpo.visualize import browse_images

browse_images(training_images)
```

## Build AAM

```python
from menpo.fitmultilevel.aam import AAMBuilder
from menpo.transform import ThinPlateSplines

def norm_int(image):
    norm_image = image.copy()
    norm_image.normalize_norm_inplace()
    return norm_image

aam_builder = AAMBuilder(normalization_diagonal=100, 
                         feature_type=None,
                         transform=ThinPlateSplines,
                         downscale=1.2)

aam = aam_builder.build(training_images, 
                        group='stickmen_pose', 
                        verbose=True)
```

```python
print aam
```

```python
from menpo.visualize import visualize_aam

visualize_aam(aam)
```

## Build Patch-Based AAM

```python
from menpo.fitmultilevel.aam import PatchBasedAAMBuilder

pb_aam_builder = PatchBasedAAMBuilder(normalization_diagonal=100, 
                                      patch_shape=(24, 24),
                                      feature_type='hog',
                                      downscale=1.2)
pb_aam = pb_aam_builder.build(training_images, 
                              group='stickmen_pose', 
                              verbose=True)
```

```python
print pb_aam
```

```python
from menpo.visualize import visualize_aam

visualize_aam(pb_aam)
```

## Load test data

```python
import menpo.io as mio
from menpo.landmark import labeller, ibug_face_68, ibug_face_68_trimesh

test_images = []
for i in mio.import_images('/data/PhD/DataBases/faces/lfpw/testset/', 
                           max_images=None, verbose=True):
    labeller(i, 'PTS', ibug_face_68)
    labeller(i, 'PTS', ibug_face_68_trimesh)
    i.crop_to_landmarks_proportion_inplace(0.5, group='ibug_face_68')
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
    test_images.append(i)
```

```python
from menpo.visualize import browse_images

browse_images(test_images)
```

## Build Fitters

```python
from menpo.fitmultilevel.aam import LucasKanadeAAMFitter
                                              
fitter = LucasKanadeAAMFitter(aam, n_shape=[3, 4, 5], n_appearance=50)
print fitter
```

```python
from menpo.fitmultilevel.aam import LucasKanadeAAMFitter
                                              
pb_fitter = LucasKanadeAAMFitter(pb_aam, n_shape=[3, 6, 12], n_appearance=50)
print pb_fitter
```

## Fit test data

```python
fitting_results = []
pb_fitting_results = []

for j, i in enumerate(training_images):
    gt_s = i.landmarks['stickmen_pose'].lms
    s = pb_fitter.perturb_shape(gt_s, noise_std=0.0, rotation=True)
    
    fr = fitter.fit(i, s, gt_shape=gt_s)
    pb_fr = pb_fitter.fit(i, s, gt_shape=gt_s)
    
    fitting_results.append(fr)
    pb_fitting_results.append(pb_fr)
    
    print 'Image: ', j
    print 'aam-fitter:', fr
    print 'pb-aam-fitter:', pb_fr
```

## Compare CED graphs

```python
initial_errors = [fr.initial_error() for fr in fitting_results]

final_errors = [fr.final_error() for fr in fitting_results]
pb_final_errors = [pb_fr.final_error() for pb_fr in pb_fitting_results]
```

```python
from menpo.visualize import plot_ced

plot_ced([final_errors, pb_final_errors], initial_errors=initial_errors)
```

```python
fitted_images = [fr.fitted_image for fr in fitting_results]
```

```python
from menpo.visualize import browse_fitted_images

browse_fitted_images(fitted_images)
```
