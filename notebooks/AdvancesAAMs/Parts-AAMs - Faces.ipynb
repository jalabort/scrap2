## Load training data

```python
import menpo.io as mio
from menpo.landmark import labeller, ibug_face_68, ibug_face_68_trimesh

training_images = []
for i in mio.import_images('/data/PhD/DataBases/faces/lfpw/trainset/', 
                           max_images=None, verbose=True):
    labeller(i, 'PTS', ibug_face_68)
    labeller(i, 'PTS', ibug_face_68_trimesh)
    i.crop_to_landmarks_proportion_inplace(0.5, group='ibug_face_68')
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
    training_images.append(i)
```

```python
from menpo.visualize import browse_images

browse_images(training_images)
```

## Build Parts-AAM

```python
from skimage.feature import daisy

def daisy_features(image):
    pixels = daisy(image.pixels[..., -1], step=1, rings=1, 
                   radius=4, histograms=0)
    img = Image(pixels)
    img.landmarks = image.landmarks
    return img
```

```python
from menpo.advanced_aam_paper import PartsAAMBuilder
from menpo.image import Image

builder = PartsAAMBuilder(
    parts_shape=(17, 17), feature_type=None,
    diagonal=200, scale_shapes=False, scales=[1, .5])
                                      
aam = builder.build(
    training_images, group='ibug_face_68', verbose=True)
```

## Load test data

```python
import menpo.io as mio
from menpo.landmark import labeller, ibug_face_68, ibug_face_68_trimesh

test_images = []
for i in mio.import_images('/data/PhD/DataBases/faces/lfpw/testset/', 
                           max_images=10, verbose=True):
    labeller(i, 'PTS', ibug_face_68)
    i.crop_to_landmarks_proportion_inplace(0.5, group='ibug_face_68')
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
    test_images.append(i)
```

```python
from menpo.visualize import browse_images

browse_images(test_images)
```

## Build Fitters

```python
from menpo.advanced_aam_paper import LucasKanadePartsAAMFitter
from menpo.advanced_aam_paper.algorithm import BFC_GN, AIC_GN

parts_mask = np.require(np.zeros((17, 17)), dtype=np.bool)
parts_mask[2::4,2::4] = True
                                              
fitter = LucasKanadePartsAAMFitter(
    aam, algorithm=AIC_GN, n_shape=[3, 12], n_appearance=[50, 200], mask=parts_mask)

print fitter
```

```python
imshow(parts_mask)
```

## Fit test data

```python
fitter_results = []

for j, i in enumerate(test_images):
    
    gt_s = i.landmarks['ibug_face_68'].lms
    s = fitter.perturb_shape(gt_s, noise_std=0.02)
    
    fr = fitter.fit(i, s, gt_shape=gt_s)
    
    fitter_results.append(fr)
    
    print 'Image: ', j
    print 'linear-aam-fitter:' 
    print fr
```

## Compare CED graphs

```python
initial_errors = [fr.initial_error() for fr in fitter_results]

final_errors = [fr.final_error() for fr in fitter_results]
```

```python
from menpo.visualize import plot_ced

plot_ced(final_errors, initial_errors=initial_errors)
```
