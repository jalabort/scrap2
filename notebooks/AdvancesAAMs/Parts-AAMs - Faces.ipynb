## Load training data

```python
import menpo.io as mio
from menpo.landmark import labeller, ibug_face_49
from menpo.aam.image import Image

training_images = []
for i in mio.import_images('/data/PhD/DataBases/faces/lfpw/trainset/', 
                           max_images=None, verbose=True):
    
    new_i = Image(np.rollaxis(i.pixels, -1))
    new_i.landmarks = i.landmarks
    
    labeller(new_i, 'PTS', ibug_face_49)
    new_i.crop_to_landmarks_proportion_inplace(0.5, group='PTS')
    if new_i.n_channels == 3:
        new_i = new_i.as_greyscale(mode='average')
    training_images.append(new_i)
```

```python
from menpo.visualize import browse_images

browse_images(training_images)
```

```python
from skimage.feature import daisy
from cyvlfeat.sift.dsift import dsift
from menpo.aam.image import Image

def daisy_features(image):
    pixels = daisy(image.pixels[0, ...], step=1, rings=1, 
                   radius=3, histograms=0)
    img = Image(np.rollaxis(pixels, -1))
    img.landmarks = image.landmarks
    return img

def dense_sift_features(image):
    _, pixels = dsift(np.rot90(image.pixels[0, ..., ::-1]), 
                      fast=True, window_size=17, geometry=(1,1,8))
    pixels_shape = (-1,) + (image.shape[0], image.shape[1])
    img = Image(np.require(pixels.reshape(pixels_shape), dtype=np.double))
    img.landmarks = image.landmarks
    return img
```

```python
from menpo.aam import PartsAAMBuilder

builder = PartsAAMBuilder(
    parts_shape=(17, 17), feature_type=dense_sift_features,
    diagonal=150, scale_shapes=False, scales=[1, .5])
                                      
aam = builder.build(
    training_images, group='ibug_face_49', verbose=True)
```

## Load test data

```python
import menpo.io as mio
from menpo.landmark import labeller, ibug_face_49

test_images = []
for i in mio.import_images('/data/PhD/DataBases/faces/lfpw/testset/', 
                           max_images=None, verbose=True):
    
    new_i = Image(np.rollaxis(i.pixels, -1))
    new_i.landmarks = i.landmarks
    
    labeller(new_i, 'PTS', ibug_face_49)
    new_i.crop_to_landmarks_proportion_inplace(0.5, group='PTS')
    if new_i.n_channels == 3:
        new_i = new_i.as_greyscale(mode='luminosity')
    test_images.append(new_i)
```

```python
from menpo.visualize import browse_images

browse_images(test_images)
```

## Build Fitters

```python
from menpo.aam import LucasKanadePartsAAMFitter
from menpo.aam.algorithm import BFC_GN, BSC_ESM, BBC_ESM, AIC_GN, ASC_ESM, ABC_ESM

parts_mask = np.require(np.zeros((17, 17)), dtype=np.bool)
parts_mask[2::4, 2::4] = True
                                              
fitter = LucasKanadePartsAAMFitter(
    aam, algorithm=BSC_ESM, n_shape=[3, 10], n_appearance=[10, 50], mask=parts_mask)

print fitter
```

```python
imshow(parts_mask)
```

## Fit test data

```python
fitter_results = []

for j, i in enumerate(test_images):
    
    gt_s = i.landmarks['ibug_face_49'].lms
    s = fitter.perturb_shape(gt_s, noise_std=0.04)
    
    fr = fitter.fit(i, s, gt_shape=gt_s, max_iters=5, prior=False)
    
    fitter_results.append(fr)
    
    print 'Image: ', j
    print 'linear-aam-fitter:' 
    print fr
```

## Compare CED graphs

```python
initial_errors = [fr.initial_error() for fr in fitter_results]

final_errors = [fr.final_error() for fr in fitter_results]
```
