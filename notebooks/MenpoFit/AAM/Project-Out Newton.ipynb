```python
%matplotlib inline
%pylab inline

from functools import partial

import menpo.io as mio
from menpo.feature import fast_dsift, double_igo

from menpo.landmark import labeller, ibug_face_66
from menpo.visualize import visualize_images, print_dynamic, plot_graph

from menpofit.aam import (
    HolisticAAM, LucasKanadeAAMFitter, 
    holistic_sampling_from_scale,  holistic_sampling_from_step)
from menpofit.aam.algorithm.lk_alabort_arxiv2015 import (
    ProjectOutForwardNewton, 
    ProjectOutInverseNewton,
    ProjectOutAsymmetricNewton,
    ProjectOutBidirectionalNewtonCombined,
    ProjectOutBidirectionalNewtonSimultaneous,
    ProjectOutBidirectionalNewtonAlternated)
from menpofit.visualize import visualize_fitting_result, plot_ced
```

## Load training data

```python
training_images = []
for i in mio.import_images('/Users/joan/PhD/DataBases/faces/lfpw/trainset/', 
                           verbose=True, max_images=None):
    i.rescale_landmarks_to_diagonal_range(200)
    i = i.crop_to_landmarks_proportion(0.5)
    labeller(i, 'PTS', ibug_face_66)
    if i.n_channels == 3:
        i = i.as_greyscale(mode='average')
    training_images.append(i)
```

```python
visualize_images(training_images)
```

## Load test data

```python
test_images = []
for i in mio.import_images('/Users/joan/PhD/DataBases/faces/lfpw/testset/', 
                           verbose=True, max_images=10):    
    i.rescale_landmarks_to_diagonal_range(200)
    i = i.crop_to_landmarks_proportion(0.5)
    labeller(i, 'PTS', ibug_face_66)
    if i.n_channels == 3:
        i = i.as_greyscale(mode='average')
    test_images.append(i)
```

```python
visualize_images(test_images)
```

## Build Active Appearance Models

```python
features = fast_dsift
diagonal = 200
scales = (0.5, 1)
max_shape_components= 25
max_appearance_components = None
```

#### AAM

```python
aam = HolisticAAM(
    training_images, 
    group='ibug_face_66', 
    verbose=True,
    features=features,
    diagonal=diagonal,
    scales=scales,
    max_shape_components=max_shape_components,
    max_appearance_components=max_appearance_components)
```

## Fit Active Appearance Models

```python
sampling_step = [16, 16]
n_shape = [3, 12]
n_appearance = [50, 100]
l = 0.5
```

#### AAM Fitter

```python
fitter_1 = LucasKanadeAAMFitter(
    aam, 
    lk_algorithm_cls=partial(ProjectOutForwardNewton, l=l),
    n_shape=n_shape, 
    n_appearance=n_appearance,
    sampling=sampling_step)

fitter_2 = LucasKanadeAAMFitter(
    aam, 
    lk_algorithm_cls=partial(ProjectOutInverseNewton, l=l),
    n_shape=n_shape, 
    n_appearance=n_appearance,
    sampling=sampling_step)

fitter_3 = LucasKanadeAAMFitter(
    aam, 
    lk_algorithm_cls=partial(ProjectOutAsymmetricNewton, l=l), 
    n_shape=n_shape, 
    n_appearance=n_appearance,
    sampling=sampling_step)

fitter_4 = LucasKanadeAAMFitter(
    aam, 
    lk_algorithm_cls=partial(ProjectOutBidirectionalNewtonCombined, l=l), 
    n_shape=n_shape, 
    n_appearance=n_appearance,
    sampling=sampling_step)

fitter_5 = LucasKanadeAAMFitter(
    aam, 
    lk_algorithm_cls=partial(ProjectOutBidirectionalNewtonSimultaneous, l=l),
    n_shape=n_shape, 
    n_appearance=n_appearance,
    sampling=sampling_step)

fitter_6 = LucasKanadeAAMFitter(
    aam, 
    lk_algorithm_cls=partial(ProjectOutBidirectionalNewtonAlternated, l=l),
    n_shape=n_shape, 
    n_appearance=n_appearance,
    sampling=sampling_step)
```

#### Fitting loop

```python
fitters = [fitter_1,
           fitter_2,
           fitter_3,
           fitter_4,
           fitter_5,
           fitter_6]

n_fitters = len(fitters)
legend_entries=['ini', 'f', 'i', 'a', 'b_c', 'b_s', 'b_a']

fitters_results = []

for j, i in enumerate(test_images):
    
    np.random.seed(j)
    
    gt_s = i.landmarks['ibug_face_66'].lms
    s = fitters[0].noisy_shape_from_shape(gt_s, noise_percentage=0.02)

    for fitter in fitters:
        
        fr = fitter.fit_from_shape(i, s, gt_shape=gt_s, max_iters=20) 
        fr.downscale = 0.5
        fitters_results.append(fr)
    
        print_dynamic(
            'Image: {} - Initial error: {} - Final error: {}'
            .format(j, fr.initial_error(), fr.final_error()))
```

## Visualize Results

```python
visualize_fitting_result(fitters_results)
```
