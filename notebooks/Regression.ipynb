### 1.1. Load data

This section loads all the images and corresponding annotations of a particular facial dataset (e.g. LFPW). The original 1-based annotations are corrected to be 0-based and the original images are cropt around each face discarting unecessary background information.

---

Notes: 

 * `RGBImage` and `IntensityImage` are, by default, loaded as `Double`. It might be better to load them as `Integer` and explicitely transforming them into `Double` after the Region of Interest (RoI) (e.g. face) has been cropt out (even then, they could be kept `Integer`; we could delay tranforming them into `Double` until a rescaling, warping or feature computation operation is required). This way much larger datasets could be loaded in memory at once and no precision would be lost.

### 1.2. Compute mean shape

The next step consists of computing the mean shape given by all ground truth annotations. In the next seccion, this mean shape will be used to map all images onto a common *scale* space.

```python
from pybug.shape import PointCloud
from pybug.groupalign import GeneralizedProcrustesAnalysis

# pull out the shape information
shapes = [img.landmarks['PTS'].lms for img in images]
# center shapes
cent_shapes = [Translation(-s.centre).apply(s) for s in shapes]
```

```python
# perform procrustes analysis
gpa = GeneralizedProcrustesAnalysis(cent_shapes)
aligned_shapes = [p.aligned_source for p in gpa.transforms]
```

```python
# compute the mean shape
mean_shape = PointCloud(np.mean([p.points for p in aligned_shapes], axis=0))
```

## 2. Training

In this section we train the cascade of linear regressors for facial alignment.

### 2.1. Extract Patches

We start by defining a function that given an annotated image allows us to extract localized image information around each landmark.

---

Notes:

 * Ideally this function should be fully integrated in the `PyBug` image package; perhaps as a method in `MaskedNDImage`.
 * More advance methods for optaining local patches could be implemented (e.g. sampling taking into account possible in-plane rotations of the face).
 * It would also be interesting to develop methods for visualing the patches in the spatial domain of defined by the original image, perhaps using a mask that allows us to show only the image patch around each landmark.

```python
def compute_patch_descriptors(img, shape, patch_size, feature_function, **kwarg):
    
    patch_length = patch_size[0] * patch_size[1] * img.n_channels
    patch_half_size = patch_size / 2
    patch_descriptors = []
    
    for i, point in enumerate(shape.points):
        start = np.round(point - patch_half_size).astype(int)
        finish = np.round(point + patch_half_size).astype(int) 
        y, x = np.mgrid[start[0]:finish[0],start[1]:finish[1]]
        y[y > img.shape[0]] = img.shape[0]
        x[x > img.shape[1]] = img.shape[1]
        y[y > 0] = 0
        x[x > 1] = 0
        patch = img.pixels[y.flatten(), x.flatten()]
        patch = patch.reshape(patch_size)
        patch = feature_function(patch[..., None], **kwarg)[0]
        patch_descriptors.append(patch.flatten())
        
    return patch_descriptors
```

```python
def build_patch_mask(img, shape, patch_size):

    mask = np.zeros(img.shape)
    patch_half_size = patch_size / 2
    
    for i, point in enumerate(shape.points):
        start = np.round(point - patch_half_size).astype(int)
        finish = np.round(point + patch_half_size).astype(int) 
        x, y = np.mgrid[start[0]:finish[0],start[1]:finish[1]]
        mask[x.flatten(), y.flatten()] = True
        
    return mask
```

### 2.2. Perturb Shapes

```python
from pybug.transform import SimilarityTransform

def perturb_shape(shape, mean_shape, n_perturbations):
    
    # initilaize the list of pertubed shapes
    perturbed_shapes = []
    
    # define standard deviation for each similarity parameters
    s_std = 0.02
    r_std = 0.02
    t_std = 2
    
    for i in range(0,n_perturbations):
        # sample scale from a Gaussian distribution with mean = 0 and std = s_std
        s = s_std * np.random.standard_normal()
        # sample rotation from a Gaussian distribution with mean = 0 and std = r_std
        r = r_std * np.random.standard_normal()
        # sample translation in both x and y from a Gaussian distribution with mean = 0 and std = t_std
        tx = t_std * np.random.standard_normal()
        ty = t_std * np.random.standard_normal()
        
        # compute the true similarity transform between the mean shape and the ground truth shape
        transform = SimilarityTransform.align(mean_shape, shape)
        # perturbed the parameters of the previous similarity transform using the previous s, r, tx and ty
        perturbed_param = transform.as_vector() + np.array([s, r, tx, ty])
        # use the previous parameters to obtain the perturbed similarity transform
        perturbed_transform = SimilarityTransform.from_vector(perturbed_param)
        # apply the previous similarity transform to the mean shape and append it to the list of perturbed shapes
        perturbed_shapes.append(perturbed_transform.apply(mean_shape))
    
    # return the list of perturbed shapes
    return perturbed_shapes
```

### 2.3. Simple Linear Regression

```python
from pybug.decomposition.base import _eigenvalue_decomposition

def linear_regression(images, current_shapes, patch_size, feature_function, **kwarg):

    n_points = current_shapes[0][0].n_points * current_shapes[0][0].n_dims
    n_features = 9792 + 1
    
    # initialize
    features = []
    a = np.zeros((n_points, n_features))
    cov = np.zeros((n_features, n_features))
    count = 0
    
    for img, shapes in zip(images, current_shapes):
        
        features_img = []
        
        for s in shapes:
            
            fi = np.array(compute_patch_descriptors(img, s, patch_size, feature_function, **kwarg)).flatten()
            fi = np.hstack((fi, 1))
            features_img.append(fi)
            cov = cov + np.dot(fi[..., None], fi[..., None].T)
            delta = img.landmarks['PTS'].lms.points - s.points
            a = a + np.dot(delta.flatten()[..., None], fi[None, ...])
        
        features.append(features_img)
        
        print count
        count += 1
        
    shapes = [img.landmarks['PTS'].lms.points.flatten() for img in images]
            
    # make sure the covariance matrix is symmetric; this is pretty stupid to be fair...
    cov = (cov + cov.T) / 2
    # eigendecompose the covariance matrix
    pos_eigenvectors, pos_eigenvalues = _eigenvalue_decomposition(cov)
    # and approximate its inverse by: cov^{-1} \approx U * L^{1} * U
    inv_cov =  np.dot(np.dot(pos_eigenvectors, np.diag(1 / pos_eigenvalues)), pos_eigenvectors.T)
    
    # compute the descent direction
    w = np.dot(a, inv_cov)
    
    # update the current shapes
    current_shapes = [[s + np.dot(w, f[..., None]) for f in fi] 
                      for s, fi in zip(shapes, features)]
    
    # compute current rmse
    error = 0
    for s, current in zip(shapes, current_shapes):
        for c in current:
            error += (s - c) ** 2
    rmse = np.sqrt(np.mean(error))
    
    # print current rmse
    print 'The current RMSE over the training set is: {}'.format(rmse)
    
    # return w, cov and current_shapes
    return w, cov, current_shapes
```
