```python
import menpo.io as mio
from menpo.landmark import labeller, ibug_face_66
from aam.image import Image

training_images = []
for i in mio.import_images('/data/PhD/DataBases/faces/lfpw/trainset/', 
                           max_images=100, verbose=True):
    
    new_i = Image(np.rollaxis(i.pixels, -1))
    new_i.landmarks = i.landmarks
    
    labeller(new_i, 'PTS', ibug_face_66)
    new_i.crop_to_landmarks_proportion_inplace(0.5, group='ibug_face_66')
    if new_i.n_channels == 3:
        new_i = new_i.as_greyscale(mode='average')
    training_images.append(new_i)
```

```python
sample_offsets = np.array((0, 0))[..., None] #np.meshgrid(range(-4, 5, 4), range(-4, 5, 4)) #  #
sample_offsets = np.asarray([sample_offsets[0].flatten(), sample_offsets[1].flatten()]).T 
n_s = sample_offsets.shape[0]
print n_s
```

```python
from aam.cython import extract_patches_cython

patch_shape = (17, 17)
lim_y = patch_shape[0] / 2
lim_x = patch_shape[1] / 2
training_parts_images = [extract_patches_cython(i.pixels, i.landmarks['ibug_face_66'].lms.points, 
                                                2*np.array(patch_shape), sample_offsets) [:, :, :, :, :]
                         for i in training_images]
```

```python
import scipy

mask = sum(meshgrid(scipy.signal.cosine(34), scipy.signal.cosine(34)), axis=0)
```

```python
expanded_images = []
for i in training_parts_images:
    aux = np.zeros(i.shape[:-2] + (2*patch_shape[0], 2*patch_shape[1]))
    for j in xrange(i.shape[1]):
        #aux[:, j, :, :, :] = np.zeros((n_channels, n_s, 2*patch_shape[0], 2*patch_shape[1]))
        #aux[:, j, :, -lim_y:, -lim_x:] = i[:, j, :, :lim_y, :lim_x]
        #aux[:, j, :, :lim_y+1, -lim_x:] = i[:, j, :, lim_y:, :lim_x]
        #aux[:, j, :, -lim_y:, :lim_x+1] = i[:, j, :, :lim_y, lim_x:]
        #aux[:, j, :, :lim_y+1, :lim_x+1] = i[:, j, :, lim_y:, lim_x:]
        for c in xrange(n_channels):
            for k in xrange(n_s):
                #aux[c, j, k, ...] = np.fft.fftshift(mask * aux[c, j, k, ...])
                aux[c, j, k, ...] = mask * i[c, j, k, ...]
    expanded_images.append(aux)
```

```python
parts_images_hat = []
pixels_hat = np.zeros_like(expanded_images[0], dtype=np.complex)
for i in expanded_images:
    for j in xrange(training_images[0].landmarks['ibug_face_66'].lms.n_points):
        for c in xrange(n_channels):
            for k in xrange(n_s):
                pixels_hat[c, j, k, :, :] = np.fft.fft2(i[c, j, k, :, :])
    parts_images_hat.append(Image(pixels_hat))
```

```python
from scipy.stats import multivariate_normal
from menpo.fitmultilevel.functions import build_sampling_grid

# Build the sampling grid associated to the patch shape
sampling_grid = build_sampling_grid(patch_shape)
# Define the 2-dimensional gaussian distribution
mean = np.zeros(2)
covariance = 3
mvn = multivariate_normal(mean=mean, cov=covariance)
# Compute Gaussian-KDE grid
kernel_grid = mvn.pdf(sampling_grid)

kernel_grids = []
for sam_off in sample_offsets:
    kernel_grids.append(mvn.pdf(sampling_grid + sam_off))
```

```python
ks = []
for kernel_grid in kernel_grids:
    aux = np.zeros((2*patch_shape[0], 2*patch_shape[1]))
    aux[-lim_y:, -lim_x:] = kernel_grid[:lim_y, :lim_x]
    aux[:lim_y+1, -lim_x:] = kernel_grid[lim_y:, :lim_x]
    aux[-lim_y:, :lim_x+1] = kernel_grid[:lim_y, lim_x:]
    aux[:lim_y+1, :lim_x+1] = kernel_grid[lim_y:, lim_x:]
    aux = np.fft.fftshift(aux)
    ks.append(aux)
```

```python
imshow(ks[0])
```

```python
y_hat = []
for k in ks:
    y_hat.append(np.fft.fft2(k))

imshow(np.abs(np.fft.fftshift(y_hat[0])))
```

```python
l = 0
part = 0
```

```python
h = np.zeros((n_channels,) + (2*patch_shape[0], 2*patch_shape[1]), dtype=np.complex)

for j in xrange(2*patch_shape[0]):
    for k in xrange(2*patch_shape[1]):
        H_hat = np.zeros((n_channels, n_channels), dtype=np.complex)
        J_hat = np.zeros((n_channels,), dtype=np.complex)
        for xi_hat in parts_images_hat:
            xi_hat = xi_hat.pixels
            for s, yi_hat in enumerate(y_hat):
                H_hat = H_hat + np.conj(xi_hat[:, part, s, j, k])[:][..., None].dot(np.conj(xi_hat[:, part, s, j, k])[:][None, ...])
                J_hat = J_hat + np.conj(xi_hat[:, part, s, j, k])[:] * yi_hat[j, k]
        H_hat = H_hat + l * np.eye(H_hat.shape[0])
        h[..., j, k] = np.linalg.solve(H_hat, J_hat)
```
