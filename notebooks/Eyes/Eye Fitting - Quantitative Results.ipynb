## 2 Build open/close eye classifier

### 2.1 Generate open samples

```python
from menpo.shape import PointCloud
from menpo.image.feature.features import hog, igo
from menpo.fitmultilevel.functions import extract_local_patches_fast, mean_pointcloud, noisy_align

patch_shape = (16, 16)
feature_patch_length = 2*16*16
open_ind = np.array([0, 6, 22, 26, 30, 32, 34])
close_ind = np.hstack((np.array([0, 7]), np.arange(12, 17)))
feature_sample_length = len(open_ind) * feature_patch_length

open_shape_mean = mean_pointcloud([i.landmarks['PTS'].lms for i in training_open_right_eye])
close_shape_mean =  mean_pointcloud([i.landmarks['PTS'].lms for i in training_close_right_eye])

correct_samples = np.zeros((10*(len(training_open_right_eye) + len(training_close_right_eye)), feature_sample_length))
wrong_samples = np.zeros((10*(len(training_open_right_eye) + len(training_close_right_eye)), feature_sample_length))

count_correct = 0
count_wrong = 0
for k, i in enumerate(training_open_right_eye):
    for _ in range(10):
        correct_shape = i.landmarks['PTS'].lms
        correct_shape = noisy_align(correct_shape, correct_shape, noise_std=0.02).apply(correct_shape)
        correct_centers = PointCloud(correct_shape.points[open_ind, :])

        patches = extract_local_patches_fast(i, correct_centers, patch_shape)
        features = np.zeros((len(open_ind), feature_patch_length))
        for j, patch in enumerate(patches):
            features[j, ...] =igo(patch).ravel()
        correct_samples[count_correct, ...] = features.ravel()
        
        count_correct += 1
    
    for _ in range(10):
        wrong_shape = noisy_align(PointCloud(close_shape_mean.points[:12]), PointCloud(correct_shape.points[:12]),
                                  noise_std=0.02).apply(close_shape_mean)
        wrong_centers = PointCloud(wrong_shape.points[close_ind, :])

        patches = extract_local_patches_fast(i, wrong_centers, patch_shape)
        features = np.zeros((len(close_ind), feature_patch_length))
        for j, patch in enumerate(patches):
            features[j, ...] = igo(patch).ravel()
        wrong_samples[count_wrong, ...] = features.ravel()
        
        count_wrong += 1
```

### 2.2 Generate close samples

```python
from menpo.shape import PointCloud
from menpo.image.feature.features import hog, igo
from menpo.fitmultilevel.functions import extract_local_patches_fast, mean_pointcloud, noisy_align

close_ind = np.hstack((np.array([0, 6]), np.arange(12, 17)))
feature_patch_length = 2*16*16
feature_sample_length = len(open_ind) * feature_patch_length

for k, i in enumerate(training_close_right_eye):
    for _ in range(10):
        correct_shape = i.landmarks['PTS'].lms
        correct_shape = noisy_align(correct_shape, correct_shape, noise_std=0.02).apply(correct_shape)
        correct_centers = PointCloud(correct_shape.points[close_ind, :])

        patches = extract_local_patches_fast(i, correct_centers, patch_shape)
        features = np.zeros((len(close_ind), feature_patch_length))
        for j, patch in enumerate(patches):
            features[j, ...] = igo(patch).ravel()
        correct_samples[count_correct, ...] = features.ravel()

        count_correct += 1
    
    for _ in range(10):
        wrong_shape = noisy_align(PointCloud(open_shape_mean.points[:12]), PointCloud(correct_shape.points[:12]),
                                  noise_std=0.02).apply(open_shape_mean)
        wrong_centers = PointCloud(wrong_shape.points[open_ind, :])

        patches = extract_local_patches_fast(i, wrong_centers, patch_shape)
        features = np.zeros((len(close_ind), feature_patch_length))
        for j, patch in enumerate(patches):
            features[j, ...] = igo(patch).ravel()
        wrong_samples[count_wrong, ...] = features.ravel()
        
        count_wrong += 1
```

```python
def extract_hog_features(img, centers, patch_shape=(16, 16)):
    patches = extract_local_patches_fast(img, centers, patch_shape)
    features = np.zeros((centers.n_points, 2*16*16))
    for j, patch in enumerate(patches):
        features[j, ...] = igo(patch).ravel()

    return features.ravel()
```

### 2.3 Define and train open/close eye classifier

Define a general SVM plus Logistic Regression classifier:

```python
from sklearn import svm
from sklearn import linear_model

def linear_svm_lr(X, t):
    r"""
    Binary classifier that combines Linear Support Vector Machines and
    Logistic Regression.
    """
    clf1 = svm.LinearSVC(class_weight='auto')
    clf1.fit(X, t)
    t1 = clf1.decision_function(X)
    clf2 = linear_model.LogisticRegression(class_weight='auto')
    clf2.fit(t1[..., None], t)

    def predict(x):
        t1_pred = clf1.decision_function(x)
        return clf2.predict_proba(t1_pred[..., None])

    return predict
```

Train it using the previous open and close samples:

```python
X = np.vstack((correct_samples, wrong_samples))
t = np.hstack((np.zeros(correct_samples.shape[0]), np.ones(wrong_samples.shape[0])))

classifier = linear_svm_lr(X, t)
```
