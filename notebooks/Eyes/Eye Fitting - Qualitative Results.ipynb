```python
import matplotlib.pylab as plt
from IPython.html.widgets import interact
import numpy as np


def browse_final_eye_images(final_images, plot_size=(18, 18), linewidth=2, color='r', marker='s'):
    r"""
    Allows browsing through a list of final images, obtained by calling the
    property `final_image` on a :map:`FittingResult` or subclass, using a
    simple sliders.

    Parameters
    -----------
    final_images : `list` of :map:`Images` or subclass
        The list of final images to be displayed.

    plot_size : (`int`, `int`), optional
        The size of the plotted figures.
    """
    n_images = len(final_images)
    size = plot_size

    def view_final_image(image):
        plot_siz = (18, 18)

        if 'ground' in final_images[image].landmarks:
            if size != (18, 18):
                plot_siz = (18, 18)
            plt.subplot(1, 3, 1)
            final_images[image].view()
            plot_eye_landmarks(
                final_images[image].landmarks['initial'].lms.points, linewidth, color, marker)
            plt.subplot(1, 3, 2)
            final_images[image].view()
            plot_eye_landmarks(
                final_images[image].landmarks['final'].lms.points, linewidth, color, marker)
            plt.subplot(1, 3, 3)
            final_images[image].view()
            plot_eye_landmarks(
                final_images[image].landmarks['ground'].lms.points, linewidth, color, marker)
        else:
            if size != (18, 18):
                plot_siz = (12, 12)
            plt.subplot(1, 2, 1)
            final_images[image].view()
            plot_eye_landmarks(
                final_images[image].landmarks['initial'].lms.points, linewidth, color, marker)
            plt.subplot(1, 2, 2)
            final_images[image].view()
            plot_eye_landmarks(
                final_images[image].landmarks['final'].lms.points, linewidth, color, marker)
        plt.axis('off')
        plt.gcf().set_size_inches(plot_siz)

    interact(view_final_image, image=(0, n_images-1))


def plot_eye_landmarks(landmarks, w=2, c='r', m='s'):

    plt.hold(True)

    if landmarks.shape[0] == 38:
        ind = np.hstack((np.array([6]), np.arange(16, 11, -1), np.array([0]), np.arange(21, 16, -1), np.array([6])))
        plt.plot(landmarks[ind, 1], landmarks[ind, 0], color=c, linewidth=w, marker=m)
        ind = np.hstack((np.arange(22, 30), np.array([22])))
        plt.plot(landmarks[ind, 1], landmarks[ind, 0], color=c, linewidth=w, marker=m)
        ind = np.hstack((np.arange(30, 38), np.array([30])))
        plt.plot(landmarks[ind, 1], landmarks[ind, 0], color=c, linewidth=w, marker=m)

    elif landmarks.shape[0] == 17:
        ind = np.hstack((np.arange(0, 12), np.array([0])))
        plt.plot(landmarks[ind, 1], landmarks[ind, 0], color=c, linewidth=w, marker=m)
        ind = np.hstack((np.array([0]), np.arange(12, 17), np.array([6])))
        plt.plot(landmarks[ind, 1], landmarks[ind, 0], color=c, linewidth=w, marker=m)

    plt.hold(False)
```
