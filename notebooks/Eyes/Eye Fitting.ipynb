# Deformable eye tracking in-the-wild

### 1. Load open right eye images

Load open right eye annotated images:

```python
%matplotlib inline
import menpo.io as mio
from menpo.landmark import labeller, ibug_open_eye_points, ibug_open_eye_trimesh

# load training images
open_right_eye_images = []
for i in mio.import_images('/data/PhD/DataBases/eyes/Annotated/Open/Right/*', verbose=True):
    # crop image
    i.crop_to_landmarks_proportion_inplace(0.5)
    # label it
    labeller(i, 'PTS', ibug_open_eye_points)
    labeller(i, 'PTS', ibug_open_eye_trimesh)
    # convert it to grayscale if needed
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
    # append it to the list
    open_right_eye_images.append(i)
```

Load flipped open left eye annotated images (effectively transformed into open right eye annotated images):

```python
# load training images
for i in mio.import_images('/data/PhD/DataBases/eyes/Annotated_Flip/Open/Right/*', verbose=True):
    # crop image
    i.crop_to_landmarks_proportion_inplace(0.5)
    # label it
    labeller(i, 'PTS', ibug_open_eye_points)
    labeller(i, 'PTS', ibug_open_eye_trimesh)
    # convert it to grayscale if needed
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
    # append it to the list
    open_right_eye_images.append(i)
```

Divide images in training and test sets at random:

```python
from random import sample

n_images = len(open_right_eye_images)

all_indices = range(n_images)
training_indices = sample(all_indices, n_images/2)
test_indices = [x for x in all_indices if not x in training_indices]

training_open_right_eye = [open_right_eye_images[ind] for ind in training_indices]
test_open_right_eye = [open_right_eye_images[ind] for ind in test_indices]
```

Visualize training images:

```python
from menpo.visualize.widgets import browse_images_with_groups

#visualize the first image
browse_images_with_groups(training_open_right_eye)
```

Visualize test images:

```python
from menpo.visualize.widgets import browse_images_with_groups

#visualize the first image
browse_images_with_groups(test_open_right_eye)
```

## 2. Open Right Eye CLM

```python
from menpo.fitmultilevel.clm import CLMBuilder
from menpo.fitmultilevel.featurefunctions import sparse_hog

clm = CLMBuilder(feature_type=sparse_hog, normalization_diagonal=200, 
                 downscale=1.2).build(training_images, group='PTS', verbose=True)
```

## 3. Open Right Eye SDM

```python
from menpo.fitmultilevel.sdm import SDMTrainer
from menpo.fitmultilevel.featurefunctions import sparse_hog
from menpo.fit.regression.regressionfunctions import mlr_svd

sdm_fitter = SDMTrainer(regression_type=mlr_svd, regression_features=sparse_hog, normalization_diagonal=200,
                        n_levels=4, downscale=1.2, n_perturbations=20).train(training_images, group='PTS', verbose=True)
```

```python
fitting_results = []
for j, i in enumerate(training_images):
    gt_s = i.landmarks['PTS'].lms
    s = sdm_fitter.perturb_shape(gt_s)
    fr = sdm_fitter.fit(i, s, gt_shape=gt_s)
    fitting_results.append(fr)
    
    print 'Image: ', j
    print 'Initial error: ', fr.initial_error()
    print 'Final error: ', fr.final_error()
    print '\n'
```
