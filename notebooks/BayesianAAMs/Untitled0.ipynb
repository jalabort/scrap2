```python
import menpo.io as mio
from menpo.landmark import labeller, ibug_68_trimesh, ibug_49_points

# load training images
images = []
training_images = []
for i in mio.import_images('/Users/joan/PhD/DataBases/**/**/*', verbose=True):
    # crop image
    i.crop_to_landmarks_proportion_inplace(0.1)
    # label it
    labeller(i, 'PTS', ibug_68_trimesh)
    # convert it to grayscale if needed
    if i.n_channels == 3:
        g = i.as_greyscale(mode='luminosity')
    # append it to the list
    images.append(i)
    training_images.append(g)
```

```python
from menpo.fitmultilevel.aam import AAMBuilder

aam, warped_images = AAMBuilder(feature_type=None, n_levels=1).build(training_images, group='ibug_68_trimesh', verbose=True)
```
