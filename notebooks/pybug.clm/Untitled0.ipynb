```python
import numpy as np
import scipy
```

```python
from scipy.stats import multivariate_normal
```

```python
ind = np.hstack((np.arange(-8, 9)))
```

```python
patch = np.squeeze(np.asanyarray(np.meshgrid(ind, ind)))
```

```python
x = patch[0,:,:].ravel()
y = patch[1,:,:].ravel()
xy = np.vstack((x, y))
```

```python
xy.shape
```

```python
mean = np.zeros(2)
sigma = 10
cov = sigma * np.eye(2) 
grid = np.zeros(xy.T.shape[0])

for j, x in enumerate(xy.T):
    mvn = multivariate_normal(mean = x, cov = cov)
    grid[j] = mvn.pdf(mean)
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
plt.imshow(np.reshape(grid, (17, 17)))
```
