```python
import pybug.io as pio
from pybug.landmark import labeller, ibug_68_points, ibug_68_trimesh

training_images = []
# load landmarked imarges
for i in pio.import_images('/data/PhD/DataBases/lfpw/trainset/*.png'):
    i.crop_to_landmarks_proportion(0.2)
    if i.n_channels == 3:
        training_images.append(i.as_greyscale(mode='luminosity'))
    else:
        training_images.append(i)
```

```python
from sklearn import svm
from sklearn import linear_model 
from pybug.asm.base import asm_builder

def hog_closure(image):
    return image.features.hog(window_step_horizontal=4, window_step_vertical=4)

def classifier_closure(X, y):
    clf = linear_model.LogisticRegression(class_weight='auto')
    clf.fit(X, y)
    return clf
    
build_options = {'group': 'PTS', 
                 'label': 'all',
                 'interpolator': 'scipy',
                 'diagonal_range': None,
                 'patch_size': (16, 16),
                 'n_levels': 1, 
                 'downscale': 2,
                 'scaled': True,
                 'feature_type': hog_closure ,
                 'classifier_type': classifier_closure,
                 'max_shape_components': 25}
# build aam
asm = asm_builder(training_images, **build_options)
```
