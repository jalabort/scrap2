```python
import pybug.io as pio
from pybug.landmark import labeller, ibug_68_points, ibug_68_trimesh

training_images = []
# load landmarked imarges
for i in pio.import_images('/Users/joan/PhD/DataBases/lfpw/trainset/*.png', max_images=811):
    i.crop_to_landmarks_proportion(0.4)
    if i.n_channels == 3:
        training_images.append(i.as_greyscale(mode='luminosity'))
    else:
        training_images.append(i)
```

```python
# visualize the first image
%matplotlib inline
training_images[0].landmarks['PTS'].view()
```

```python
image = training_images[0]

def hog_closure(image):
    return image.features.hog(mode='sparse')

%matplotlib inline
a = hog_closure(training_images[0])
a.landmarks['PTS'].view(channels=0)
```

```python
from pybug.aam.base import aam_builder
from pybug.transform.piecewiseaffine import PiecewiseAffineTransform
from pybug.image import Image

# label images
labeller(training_images, 'PTS', ibug_68_trimesh);

# build aam
aam = aam_builder(training_images, 
                  group='PTS', 
                  trilist=training_images[0].landmarks['ibug_68_trimesh'].lms.trilist, 
                  n_levels=3, 
                  downscale=1.1, 
                  feature_type=hog_closure, 
                  max_shape_components=25, 
                  max_appearance_components=250)
```

```python
%matplotlib inline
a.build_mask_around_landmarks((5, 5), group='PTS')
a.landmarks['PTS'].view(channels=0)
```

```python
from sklearn import svm
from sklearn import linear_model 
from pybug.asm.base import asm_builder

def classifier_closure(X, y):
    clf1 = svm.LinearSVC(class_weight='auto')
    clf1.fit(X, y)
    y_pred = clf1.decision_function(X)
    clf2 = linear_model.LogisticRegression(class_weight='auto')
    clf2.fit(y_pred[..., None], y)
    return clf1, clf2

# build aam
asm = asm_builder(training_images, group='PTS', patch_size=(5, 5), n_levels=3,
                  downscale=1.1, scaled_shape_models=False, feature_type=hog_closure, 
                  classifier_type=classifier_closure, max_shape_components=25)
```

```python
del training_images
```

```python
from pybug.asam.base import ASAM

asam = ASAM(asm.shape_models, aam.appearance_models, aam.transform_cls,
            asm.classifiers, hog_closure, asm.reference_shape, asm.downscale,
            asm.patch_size, aam.interpolator)

asam.scaled_shape_models = True
```

```python
import pybug.io as pio
from pybug.landmark import labeller, ibug_68_points, ibug_68_trimesh

test_images = []
# load landmarked imarges
for i in pio.import_images('/Users/joan/PhD/DataBases/afw/*.jpg'):
    i.crop_to_landmarks_proportion(0.4)
    if i.n_channels == 3:
        test_images.append(i.as_greyscale(mode='luminosity'))
    else:
        test_images.append(i)
```

```python
from pybug.asam.fitter import RLMSLucasKanadeASAMFitter
from pybug.asam.basicfitter import RlmsProbIC, RlmsProbFC, RlmsAltIC

# initialize Lucas-Kanade aam fitting
asam_fitter = RLMSLucasKanadeASAMFitter(asam, basic_fitter_cls=RlmsAltIC,
                                        n_shape=[3, 6, 12], n_appearance=100)
```
