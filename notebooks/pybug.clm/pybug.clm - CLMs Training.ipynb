```python
import numpy as np

# TODO: document me
def build_sampling_grid(grid_size):
    r"""
    """
    patch_size = np.array(grid_size)
    patch_half_size = (np.round(patch_size / 2)).astype(int)
    start = -patch_half_size
    end = patch_half_size + 1
    return np.mgrid[start[0]:end[0], start[1]:end[1]].swapaxes(0, 2).swapaxes(0, 1)


# TODO: document me
def mask_sampling_grid(sampling_grid, mask_size=[2, 2]):
    r"""
    """
    mask_size = np.array(mask_size)

    mask = np.zeros((sampling_grid.shape[:-1]) , dtype=np.bool)
    center = np.round(np.array(mask.shape) / 2).astype(int)

    mask_size -= [1, 1]

    start = center - mask_size
    end = center + mask_size + 1
    
    mask[start[0]:end[0], start[1]:end[1]] = True
    positive = sampling_grid[mask]
    negative = sampling_grid[~mask]
    return positive, negative
```

### - Normalize face size and generate gaussian levels:

```python
from pybug.aam.functions import mean_pointcloud
    
group = 'PTS'
label = 'all'
diagonal_range = None
n_levels = 3
downscale = 2
interpolator = 'scipy'

print '- Rescaling images'
training_shapes = [i.landmarks[group][label].lms for i in training_images]
reference_shape = mean_pointcloud(training_shapes)
if diagonal_range:
    x, y = reference_shape.range()
    scale = diagonal_range / np.sqrt(x**2 + y**2)
    Scale(scale, reference_shape.n_dims).apply_inplace(reference_shape)
training_images = [i.rescale_to_reference_shape(reference_shape, group=group, 
                                                label=label, interpolator=interpolator)
                   for i in training_images]


print '- Setting gaussian pyramid generators'
generator = [i.gaussian_pyramid(n_levels=n_levels,
                                downscale=downscale)
             for i in training_images]
```

### - Generate first level

```python
from pybug.aam.functions import compute_features

def hog_closure(image):
    return image.features.hog(window_step_horizontal=4, window_step_vertical=4)

print '  - Computing feature_type'
training_images = [compute_features(g.next(), hog_closure) for g in generator]
# extract potentially rescaled shapes
training_shapes = [i.landmarks[group][label].lms for i in training_images]
```

```python
%matplotlib inline
#visualize the first image
training_images[0].landmarks['PTS'].view(channels=0)
```

```python
import numpy as np
ind = np.hstack((np.arange(-8, 0), np.arange(1,9)))
patch = np.asanyarray(np.meshgrid(ind, ind))
```

```python
from sklearn import svm
from sklearn import linear_model 

classifiers = []
for ind in np.arange(5):
    positive_labels = []
    negative_labels = []
    positive_samples = []
    negative_samples = []
    for i, s in zip(training_images, training_shapes):
        point = (np.round(s.points[ind,:])).astype(int)
        positive_sample = i.pixels[point[0], point[1], :]
        positive_samples.append(positive_sample)
        positive_labels.append(np.ones(1))
        
        selectors = patch + point[..., None, None]
        
        x = selectors[0,:,:].ravel()
        y = selectors[1,:,:].ravel()
        max_x = i.shape[0] - 1
        max_y = i.shape[1] - 1
        x[x > max_x] = max_x
        y[y > max_y] = max_y
        x[x < 0] = 0
        y[y < 0] = 0
        
        negative_sample = i.pixels[x, y, :]
        negative_samples.append(negative_sample)
        negative_labels.append(-np.ones(negative_sample.shape[0]))
    
    positive_samples = np.asanyarray(positive_samples)
    positive_labels = np.asanyarray(positive_labels).flatten()

    negative_samples = np.asanyarray(negative_samples)
    negative_samples = np.reshape(negative_samples, (-1, negative_samples.shape[-1]))
    negative_labels = np.asanyarray(negative_labels).flatten()
    
    X = np.vstack((positive_samples, negative_samples))
    y = np.hstack((positive_labels, negative_labels))
    
    svc = svm.LinearSVC(class_weight='auto', probability=True)
    svc.fit(X, y)
    classifiers.append(svc)
    #svc = svm.SVC(kernel='rbf', class_weight='auto', probability=True)
    #svc.fit(X, y)
    #classifiers.append(svc)
    #lr = linear_model.LogisticRegression(class_weight='auto', dual=True)
    #lr.fit(X, y)
    #classifiers.append(lr)
    #sdg = linear_model.SGDClassifier(class_weight='auto', loss='squared_hinge')
    #sdg.fit(X, y)
    #classifiers.append(sdg)
```
