```python
import pybug.io as pio
from pybug.landmark import labeller, ibug_68_points, ibug_68_trimesh

training_images = []
# load landmarked imarges
for i in pio.import_images('/data/PhD/DataBases/lfpw/trainset/*.png', max_images=811):
    i.crop_to_landmarks_proportion(0.4)
    if i.n_channels == 3:
        training_images.append(i.as_greyscale(mode='luminosity'))
    else:
        training_images.append(i)
```

```python
# visualize the first image
%matplotlib inline
training_images[0].landmarks['PTS'].view()
```

```python
image = training_images[0]

def hog_closure(image):
    return image.features.hog(mode='sparse')

%matplotlib inline
a = hog_closure(training_images[0])
a.build_mask_around_landmarks((5, 5))
a.landmarks['PTS'].view(channels=0)
```

```python
from sklearn import svm
from sklearn import linear_model 
from pybug.asm.base import asm_builder

def classifier_closure(X, y):
    clf1 = svm.LinearSVC(class_weight='auto')
    clf1.fit(X, y)
    y_pred = clf1.decision_function(X)
    clf2 = linear_model.LogisticRegression(class_weight='auto')
    clf2.fit(y_pred[..., None], y)
    return clf1, clf2

# build aam
asm = asm_builder(training_images, group='PTS', patch_size=(5, 5), n_levels=3,
                  downscale=1.1, scaled_shape_models=False, feature_type=hog_closure, 
                  classifier_type=classifier_closure, max_shape_components=25)
```

```python
del training_images
```

```python
import pybug.io as pio
from pybug.landmark import labeller, ibug_68_points, ibug_68_trimesh

test_images = []
# load landmarked imarges
for i in pio.import_images('/data/PhD/DataBases/afw/*.jpg'):
    i.crop_to_landmarks_proportion(0.4)
    if i.n_channels == 3:
        test_images.append(i.as_greyscale(mode='luminosity'))
    else:
        test_images.append(i)
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
plt.imshow((asm.response_image(test_images[0]).pixels[:,:,0]))
```

```python
from pybug.asm.fitter import RlmsASMFitter

asm_fitter = RlmsASMFitter(asm, n_shape=[3, 6, 12])
```

```python
%matplotlib inline
asm_fittings = asm_fitter.fit_images(test_images, group='PTS', noise_std=0.02, max_iters=50)
```

```python
%matplotlib inline
asm_fittings.plot_cumulative_error_dist(color_list=['r', 'b'], marker_list=['*', 's'])
```

```python
from copy import deepcopy
from pybug.landmark import ibug_49_points
import numpy as np
from pybug.aam.functions import _compute_me_norm
from scipy.io import savemat

final_shapes = [f.final_shape.points for f in asm_fittings] 
initial_shapes = [f.initial_shape.points for f in asm_fittings] 
gt_shapes = [f.gt_shape.points for f in asm_fittings] 

final_error = np.array([_compute_me_norm(f, gt) 
                        for (f, gt) in zip(final_shapes, gt_shapes)])
initial_error = np.array([_compute_me_norm(f, gt) 
                          for (f, gt) in zip(initial_shapes, gt_shapes)])

convergence = (np.sum(initial_error > final_error) / len(initial_error))

# set path
path =  '/data/PhD/Results/'
# set name
name = 'asm_' + 'afw_02' + '68'

savemat(path + name, {'final_error': final_error,
                      'initial_error': initial_error,
                      'convergence': convergence})
```
