## Load training data

```python
import menpo.io as mio
from menpo.landmark import labeller, ibug_face_66
from fg2015.utils import convert_from_menpo

training_images = []
for i in mio.import_images('/data/PhD/DataBases/faces/lfpw/trainset/', 
                           max_images=50, verbose=True):
    
    # convert the image from menpo Image to fg2015 Image (channels at front)
    i = convert_from_menpo(i)
    
    labeller(i, 'PTS', ibug_face_66)
    i.crop_to_landmarks_proportion_inplace(0.5, group='PTS')
    if i.n_channels == 3:
        i = i.as_greyscale(mode='average')
    training_images.append(i)
```

```python
from menpo.visualize import browse_images

browse_images(training_images)
```

## Load test data

```python
import menpo.io as mio
from menpo.landmark import labeller, ibug_face_66
from fg2015.utils import convert_from_menpo

test_images = []
for i in mio.import_images('/data/PhD/DataBases/faces/lfpw/testset/', 
                           max_images=10, verbose=True):
    
    # convert the image from menpo Image to fg2015 Image (channels at front)
    i = convert_from_menpo(i)
    
    labeller(i, 'PTS', ibug_face_66)
    i.crop_to_landmarks_proportion_inplace(0.5, group='PTS')
    if i.n_channels == 3:
        i = i.as_greyscale(mode='average')
    test_images.append(i)
```

```python
from menpo.visualize import browse_images

browse_images(test_images)
```

## Active Appearance Models

#### Build

```python
from fg2015.feature import no_op, aam_dsift
from fg2015.deformablemodel import PartsAAMBuilder

builder = PartsAAMBuilder(parts_shape=(17, 17), features=aam_dsift, diagonal=100, 
                          scale_shapes=False, scales=(1, .5))
aam = builder.build(training_images, group='ibug_face_66', verbose=True)
```

#### Test

```python
from fg2015.deformablemodel import PartsAAMFitter
from fg2015.deformablemodel.aam import AIC, PIC

sampling_mask = np.require(np.zeros((17, 17)), dtype=np.bool)
sampling_mask[1::4, 1::4] = True
                  
fitter = PartsAAMFitter(aam, algorithm_cls=AIC, n_shape=[3, 12], 
                        n_appearance=[25, 50], sampling_mask=sampling_mask)
```

```python
np.random.seed(seed=1)

fitter_results = []

for j, i in enumerate(test_images[:]):
    
    gt_s = i.landmarks['ibug_face_66'].lms
    s = fitter.perturb_shape(gt_s, noise_std=0.04)
    
    fr = fitter.fit(i, s, gt_shape=gt_s, max_iters=20)
    
    fitter_results.append(fr)
    
    print 'Image: ', j
    print fr
```

#### Plot CED

```python
initial_errors = [fr.initial_error() for fr in fitter_results]
final_errors = [fr.final_error() for fr in fitter_results]
```

```python
from menpo.visualize import plot_ced

plot_ced(final_errors, initial_errors=initial_errors)
```

#### Visualize results

```python
fitted_images = [fr.fitted_image for fr in fitter_results]
```

```python
from menpo.visualize import browse_fitted_images

browse_fitted_images(fitted_images)
```

## Constrained Local Models

```python
from fg2015.feature import no_op, aam_dsift
from fg2015.deformablemodel import CLMBuilder

offsets = np.meshgrid(range(-4, 5, 2), range(-4, 5, 2))
offsets = np.asarray([offsets[0].flatten(), offsets[1].flatten()]).T 

builder = CLMBuilder(parts_shape=(17, 17), features=aam_dsift, diagonal=200, 
                     offsets=offsets, normalize_parts=False, covariance=2, 
                     scale_shapes=False, scales=(1,))
clm = builder.build(training_images[:25], group='ibug_face_66', verbose=True)
```
